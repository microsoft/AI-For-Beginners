## ಕಾರ್ಯ: ಡಯೋಫಾಂಟೈನ್ ಸಮೀಕರಣಗಳು

> ಈ ಕಾರ್ಯ [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) ಭಾಗವಾಗಿದ್ದು, [ಈ ಪೋಸ್ಟ್](https://habr.com/post/128704/)ನಿಂದ ಪ್ರೇರಿತವಾಗಿದೆ.

ನಿಮ್ಮ ಗುರಿ ಎಂದರೆ **ಡಯೋಫಾಂಟೈನ್ ಸಮೀಕರಣ** ಅನ್ನು ಪರಿಹರಿಸುವುದು - ಅಂದರೆ ಪೂರ್ಣಾಂಕ ಮೂಲಗಳು ಮತ್ತು ಪೂರ್ಣಾಂಕ ಗುಣಾಂಕಗಳಿರುವ ಸಮೀಕರಣ. ಉದಾಹರಣೆಗೆ, ಕೆಳಗಿನ ಸಮೀಕರಣವನ್ನು ಪರಿಗಣಿಸಿ:

$$a+2b+3c+4d=30$$

ನೀವು ಈ ಸಮೀಕರಣವನ್ನು ತೃಪ್ತಿಪಡಿಸುವ ಪೂರ್ಣಾಂಕ ಮೂಲಗಳು $a$,$b$,$c$,$d\in\mathbb{N}$ ಅನ್ನು ಕಂಡುಹಿಡಿಯಬೇಕು.

ಸೂಚನೆಗಳು:
1. ಮೂಲಗಳನ್ನು [0;30] ನಡುವಿನ ಅಂತರದಲ್ಲಿ ಪರಿಗಣಿಸಬಹುದು
1. ಒಂದು ಜೀನ್ ಆಗಿ, ಮೂಲ ಮೌಲ್ಯಗಳ ಪಟ್ಟಿಯನ್ನು ಬಳಸಬಹುದು
