# # ಎಸ್ಕೇಪ್ ಮಾಡಲು ಮಾಉಂಟನ್ ಕಾರ್ ತರಬೇತಿ

[AI for Beginners Curriculum](https://github.com/microsoft/ai-for-beginners) ನಿಂದ ಪ್ರಯೋಗಾಲಯ ಕಾರ್ಯ.

ನಿಮ್ಮ ಗುರಿ RL ಏಜೆಂಟ್ ಅನ್ನು ತರಬೇತಿಮಾಡಿ OpenAI ಪರಿಸರದಲ್ಲಿ [ಮೌಂಟನ್ ಕಾರ್](https://www.gymlibrary.ml/environments/classic_control/mountain_car/) ಅನ್ನು ನಿಯಂತ್ರಿಸುವುದು.

ಪರಿಸರವನ್ನು ರಚಿಸುವುದರಿಂದ ಪ್ರಾರಂಭಿಸೋಣ:


```python
import gym
env = gym.make('MountainCar-v0')
```

ನಾವು ಯಾದೃಚ್ಛಿಕ ಪ್ರಯೋಗ ಹೇಗಿರುತ್ತದೆ ಎಂದು ನೋಡೋಣ:


```python
state = env.reset()
while True:
    env.render()
    action = env.action_space.sample()
    state, reward, done, info = env.step(action)
    if done:
        break
```

ಈಗ ನೋಟ್ಬುಕ್ ಸಂಪೂರ್ಣವಾಗಿ ನಿಮ್ಮದಾಗಿದೆ - ಈ ಸಮಸ್ಯೆಗೆ ಪಾಠದಿಂದ ಪಾಲಿಸಿ ಗ್ರೇಡಿಯಂಟ್ಸ್ ಮತ್ತು ಅಕ್ಟರ್-ಕ್ರಿಟಿಕ್ ಆಲ್ಗಾರಿಥಮ್‌ಗಳನ್ನು ಸ್ವೀಕರಿಸಲು ಮುಕ್ತವಾಗಿರಿ!
