# ಗಮನ ಯಂತ್ರಗಳು ಮತ್ತು ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್‌ಗಳು

ಪುನರಾವರ್ತಿತ ನೆಟ್‌ವರ್ಕ್‌ಗಳ ಪ್ರಮುಖ ದೋಷವೆಂದರೆ ಸರಣಿಯಲ್ಲಿನ ಎಲ್ಲಾ ಪದಗಳು ಫಲಿತಾಂಶದ ಮೇಲೆ ಸಮಾನ ಪ್ರಭಾವ ಬೀರುತ್ತವೆ. ಇದರಿಂದ ನಾಮಿತ ಘಟಕ ಗುರುತಿಸುವಿಕೆ ಮತ್ತು ಯಂತ್ರ ಅನುವಾದದಂತಹ ಸರಣಿ-ನಿಂದ-ಸರಣಿ ಕಾರ್ಯಗಳಿಗೆ ಸಾಮಾನ್ಯ LSTM ಎನ್‌ಕೋಡರ್-ಡಿಕೋಡರ್ ಮಾದರಿಗಳ ಕಾರ್ಯಕ್ಷಮತೆ ಕಡಿಮೆಯಾಗುತ್ತದೆ. ವಾಸ್ತವದಲ್ಲಿ, ಇನ್‌ಪುಟ್ ಸರಣಿಯ ನಿರ್ದಿಷ್ಟ ಪದಗಳು ಇತರ ಪದಗಳಿಗಿಂತ ಕ್ರಮಬದ್ಧ ಔಟ್‌ಪುಟ್‌ಗಳ ಮೇಲೆ ಹೆಚ್ಚು ಪ್ರಭಾವ ಬೀರುತ್ತವೆ.

ಯಂತ್ರ ಅನುವಾದದಂತಹ ಸರಣಿ-ನಿಂದ-ಸರಣಿ ಮಾದರಿಯನ್ನು ಪರಿಗಣಿಸಿ. ಇದನ್ನು ಎರಡು ಪುನರಾವರ್ತಿತ ನೆಟ್‌ವರ್ಕ್‌ಗಳ ಮೂಲಕ ಅನುಷ್ಠಾನಗೊಳಿಸಲಾಗುತ್ತದೆ,其中 ಒಂದು ನೆಟ್‌ವರ್ಕ್ (**ಎನ್‌ಕೋಡರ್**) ಇನ್‌ಪುಟ್ ಸರಣಿಯನ್ನು ಗುಪ್ತ ಸ್ಥಿತಿಗೆ ಸಂಕುಚಿತಗೊಳಿಸುತ್ತದೆ, ಮತ್ತು ಇನ್ನೊಂದು, **ಡಿಕೋಡರ್**, ಈ ಗುಪ್ತ ಸ್ಥಿತಿಯನ್ನು ಅನುವಾದಿತ ಫಲಿತಾಂಶವಾಗಿ ವಿಸ್ತರಿಸುತ್ತದೆ. ಈ ವಿಧಾನದಲ್ಲಿ ಸಮಸ್ಯೆ ಏನೆಂದರೆ ನೆಟ್‌ವರ್ಕ್‌ನ ಅಂತಿಮ ಸ್ಥಿತಿಗೆ ವಾಕ್ಯದ ಆರಂಭವನ್ನು ನೆನಪಿಡಲು ಕಷ್ಟವಾಗುತ್ತದೆ, ಇದರಿಂದ ದೀರ್ಘ ವಾಕ್ಯಗಳಲ್ಲಿ ಮಾದರಿಯ ಗುಣಮಟ್ಟ ಕೆಡುತ್ತದೆ.

**ಗಮನ ಯಂತ್ರಗಳು** RNNನ ಪ್ರತಿ ಔಟ್‌ಪುಟ್ ಭವಿಷ್ಯವಾಣಿ ಮೇಲೆ ಪ್ರತಿ ಇನ್‌ಪುಟ್ ವೆಕ್ಟರ್‌ನ ಸಾಂದರ್ಭಿಕ ಪ್ರಭಾವವನ್ನು ತೂಕ ನೀಡುವ ವಿಧಾನವನ್ನು ಒದಗಿಸುತ್ತವೆ. ಇದನ್ನು ಅನುಷ್ಠಾನಗೊಳಿಸುವ ವಿಧಾನವೆಂದರೆ ಇನ್‌ಪುಟ್ RNN ಮತ್ತು ಔಟ್‌ಪುಟ್ RNNನ ಮಧ್ಯಂತರ ಸ್ಥಿತಿಗಳ ನಡುವೆ ಶಾರ್ಟ್‌ಕಟ್‌ಗಳನ್ನು ಸೃಷ್ಟಿಸುವುದು. ಈ ರೀತಿಯಲ್ಲಿ, ಔಟ್‌ಪುಟ್ ಚಿಹ್ನೆ $y_t$ ರಚಿಸುವಾಗ, ನಾವು ಎಲ್ಲಾ ಇನ್‌ಪುಟ್ ಗುಪ್ತ ಸ್ಥಿತಿಗಳು $h_i$ಗಳನ್ನು ವಿಭಿನ್ನ ತೂಕ ಗುಣಾಂಕಗಳು $\alpha_{t,i}$ ಜೊತೆಗೆ ಪರಿಗಣಿಸುತ್ತೇವೆ.

ಗಮನ ಯಂತ್ರಗಳು ನೈಸರ್ಗಿಕ ಭಾಷಾ ಪ್ರಕ್ರಿಯೆಯಲ್ಲಿ ಇತ್ತೀಚಿನ ಅಥವಾ ಸಮೀಪದ ಇತ್ತೀಚಿನ ಅತ್ಯುತ್ತಮ ಸ್ಥಿತಿಗೆ ಬಹುಮಾನ ನೀಡುವ ಪ್ರಮುಖ ಕಾರಣಗಳಾಗಿವೆ. ಆದರೆ ಗಮನ ಸೇರಿಸುವುದರಿಂದ ಮಾದರಿಯ ಪರಿಮಾಣಗಳು ಬಹಳಷ್ಟು ಹೆಚ್ಚಾಗುತ್ತವೆ, ಇದರಿಂದ RNNಗಳೊಂದಿಗೆ ಪ್ರಮಾಣವರ್ಧನೆ ಸಮಸ್ಯೆಗಳು ಉಂಟಾಗುತ್ತವೆ. RNNಗಳನ್ನು ಪ್ರಮಾಣವರ್ಧಿಸಲು ಪ್ರಮುಖ ಅಡ್ಡಿ ಎಂದರೆ ಮಾದರಿಗಳ ಪುನರಾವರ್ತಿತ ಸ್ವಭಾವವು ತರಬೇತಿಯನ್ನು ಬ್ಯಾಚ್ ಮತ್ತು ಸಮಾಂತರಗೊಳಿಸುವುದನ್ನು ಕಷ್ಟಕರವಾಗಿಸುತ್ತದೆ. RNNನಲ್ಲಿ ಸರಣಿಯ ಪ್ರತಿ ಅಂಶವನ್ನು ಕ್ರಮಬದ್ಧವಾಗಿ ಪ್ರಕ್ರಿಯೆ ಮಾಡಬೇಕಾಗುತ್ತದೆ, ಆದ್ದರಿಂದ ಅದನ್ನು ಸುಲಭವಾಗಿ ಸಮಾಂತರಗೊಳಿಸಲಾಗುವುದಿಲ್ಲ.

ಈ ಅಡ್ಡಿಯೊಂದಿಗೆ ಗಮನ ಯಂತ್ರಗಳ ಅಳವಡಿಕೆಯಿಂದ ನಾವು ಇಂದಿನ BERT ರಿಂದ OpenGPT3 ವರೆಗೆ ಪರಿಚಿತವಾಗಿರುವ ಅತ್ಯುತ್ತಮ ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ಮಾದರಿಗಳನ್ನು ಸೃಷ್ಟಿಸಿದ್ದೇವೆ.

## ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ಮಾದರಿಗಳು

ಪ್ರತಿ ಹಿಂದಿನ ಭವಿಷ್ಯವಾಣಿಯ ಸಾಂದರ್ಭಿಕತೆಯನ್ನು ಮುಂದಿನ ಮೌಲ್ಯಮಾಪನ ಹಂತಕ್ಕೆ ಕಳುಹಿಸುವ ಬದಲು, **ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ಮಾದರಿಗಳು** **ಸ್ಥಾನಿಕ ಎನ್‌ಕೋಡಿಂಗ್‌ಗಳು** ಮತ್ತು ಗಮನವನ್ನು ಬಳಸಿಕೊಂಡು ನೀಡಲಾದ ಪಠ್ಯದ ವಿಂಡೋದಲ್ಲಿ ಇನ್‌ಪುಟ್‌ನ ಸಾಂದರ್ಭಿಕತೆಯನ್ನು ಹಿಡಿಯುತ್ತವೆ. ಕೆಳಗಿನ ಚಿತ್ರವು ಸ್ಥಾನದ ಎನ್‌ಕೋಡಿಂಗ್‌ಗಳು ಮತ್ತು ಗಮನವು ನೀಡಲಾದ ವಿಂಡೋದಲ್ಲಿ ಸಾಂದರ್ಭಿಕತೆಯನ್ನು ಹೇಗೆ ಹಿಡಿಯುತ್ತದೆ ಎಂಬುದನ್ನು ತೋರಿಸುತ್ತದೆ.

![ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ಮಾದರಿಗಳಲ್ಲಿ ಮೌಲ್ಯಮಾಪನಗಳು ಹೇಗೆ ನಡೆಯುತ್ತವೆ ಎಂಬುದನ್ನು ತೋರಿಸುವ ಅನಿಮೇಟೆಡ್ GIF](../../../../../lessons/5-NLP/18-Transformers/images/transformer-animated-explanation.gif)

ಪ್ರತಿ ಇನ್‌ಪುಟ್ ಸ್ಥಾನವನ್ನು ಪ್ರತಿ ಔಟ್‌ಪುಟ್ ಸ್ಥಾನಕ್ಕೆ ಸ್ವತಂತ್ರವಾಗಿ ನಕ್ಷೆ ಮಾಡಲಾಗುವುದರಿಂದ, ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್‌ಗಳು RNNಗಳಿಗಿಂತ ಉತ್ತಮವಾಗಿ ಸಮಾಂತರಗೊಳಿಸಬಹುದು, ಇದರಿಂದ ಬಹು ದೊಡ್ಡ ಮತ್ತು ಹೆಚ್ಚು ಅಭಿವ್ಯಕ್ತ ಭಾಷಾ ಮಾದರಿಗಳನ್ನು ಸೃಷ್ಟಿಸಲು ಸಾಧ್ಯವಾಗುತ್ತದೆ. ಪ್ರತಿ ಗಮನ ತಲೆ ಪದಗಳ ನಡುವಿನ ವಿಭಿನ್ನ ಸಂಬಂಧಗಳನ್ನು ಕಲಿಯಲು ಬಳಸಬಹುದು, ಇದು ನಂತರದ ನೈಸರ್ಗಿಕ ಭಾಷಾ ಪ್ರಕ್ರಿಯೆ ಕಾರ್ಯಗಳನ್ನು ಸುಧಾರಿಸುತ್ತದೆ.

**BERT** (Bidirectional Encoder Representations from Transformers) ಒಂದು ಬಹಳ ದೊಡ್ಡ ಬಹುಮಟ್ಟದ ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ನೆಟ್‌ವರ್ಕ್ ಆಗಿದ್ದು, *BERT-base* ಗೆ 12 ಮಟ್ಟಗಳು ಮತ್ತು *BERT-large* ಗೆ 24 ಮಟ್ಟಗಳಿವೆ. ಈ ಮಾದರಿಯನ್ನು ಮೊದಲು ದೊಡ್ಡ ಪಠ್ಯ ಸಂಗ್ರಹ (ವಿಕಿಪೀಡಿಯಾ + ಪುಸ್ತಕಗಳು) ಮೇಲೆ ಸ್ವಯಂನಿರೀಕ್ಷಿತ ತರಬೇತಿಯಲ್ಲಿ (ವಾಕ್ಯದಲ್ಲಿ ಮಸ್ಕ್ ಮಾಡಲಾದ ಪದಗಳನ್ನು ಭವಿಷ್ಯವಾಣಿ ಮಾಡುವ ಮೂಲಕ) ಪೂರ್ವ-ತರಬೇತಿ ಮಾಡಲಾಗುತ್ತದೆ. ಪೂರ್ವ-ತರಬೇತಿಯ ಸಮಯದಲ್ಲಿ ಮಾದರಿ ಭಾಷಾ ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವ ಮಹತ್ವಪೂರ್ಣ ಮಟ್ಟವನ್ನು ಶೋಷಿಸುತ್ತದೆ, ಇದನ್ನು ನಂತರ ಇತರ ಡೇಟಾಸೆಟ್‌ಗಳೊಂದಿಗೆ ಸೂಕ್ಷ್ಮ-ಸಂಯೋಜನೆಯ ಮೂಲಕ ಉಪಯೋಗಿಸಬಹುದು. ಈ ಪ್ರಕ್ರಿಯೆಯನ್ನು **ಟ್ರಾನ್ಸ್‌ಫರ್ ಲರ್ನಿಂಗ್** ಎಂದು ಕರೆಯುತ್ತಾರೆ.

![http://jalammar.github.io/illustrated-bert/ ನಿಂದ ಚಿತ್ರ](../../../../../translated_images/kn/jalammarBERT-language-modeling-masked-lm.34f113ea5fec4362.webp)

BERT, DistilBERT, BigBird, OpenGPT3 ಮತ್ತು ಇನ್ನಷ್ಟು ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ವಾಸ್ತುಶಿಲ್ಪಗಳ ಅನೇಕ ರೂಪಾಂತರಗಳಿವೆ, ಅವುಗಳನ್ನು ಸೂಕ್ಷ್ಮ-ಸಂಯೋಜನೆ ಮಾಡಬಹುದು. [HuggingFace ಪ್ಯಾಕೇಜ್](https://github.com/huggingface/) PyTorch ಬಳಸಿ ಈ ವಾಸ್ತುಶಿಲ್ಪಗಳ ಬಹುತೇಕ ತರಬೇತಿಗಾಗಿ ಸಂಗ್ರಹಾಲಯವನ್ನು ಒದಗಿಸುತ್ತದೆ.

ಆದರೆ, ಕೆಲವೊಮ್ಮೆ ನೀವು ನಿಮ್ಮ ಸ್ವಂತ ಮಾದರಿಗಳನ್ನು ಲೋಡ್ ಮಾಡಬೇಕಾಗಬಹುದು, ಆ ಸಂದರ್ಭದಲ್ಲಿ ನೀವು ಟೋಕನೈಜರ್‌ಗಾಗಿ ಪ್ಯಾರಾಮೀಟರ್‌ಗಳು, ಮಾದರಿ ಪ್ಯಾರಾಮೀಟರ್‌ಗಳೊಂದಿಗೆ `config.json` ಫೈಲ್, ಬೈನರಿ ತೂಕಗಳು ಮುಂತಾದ ಎಲ್ಲಾ ಸಂಬಂಧಿತ ಫೈಲ್‌ಗಳನ್ನು ಒಳಗೊಂಡಿರುವ ಡೈರೆಕ್ಟರಿಯನ್ನು ಸೂಚಿಸಬಹುದು.


```python
# To load the model from Internet repository using model name. 
# Use this if you are running from your own copy of the notebooks
bert_model = 'bert-base-uncased' 

# To load the model from the directory on disk. Use this for Microsoft Learn module, because we have
# prepared all required files for you.
bert_model = './bert'

tokenizer = transformers.BertTokenizer.from_pretrained(bert_model)

MAX_SEQ_LEN = 128
PAD_INDEX = tokenizer.convert_tokens_to_ids(tokenizer.pad_token)
UNK_INDEX = tokenizer.convert_tokens_to_ids(tokenizer.unk_token)
```

`tokenizer` ವಸ್ತುವಿನಲ್ಲಿ `encode` ಫಂಕ್ಷನ್ ಇದೆ, ಇದನ್ನು ಪಠ್ಯವನ್ನು ನೇರವಾಗಿ ಎನ್‌ಕೋಡ್ ಮಾಡಲು ಬಳಸಬಹುದು:


```python
tokenizer.encode('PyTorch is a great framework for NLP')
```

ಆಮೇಲೆ, ತರಬೇತಿ ಸಮಯದಲ್ಲಿ ಡೇಟಾವನ್ನು ಪ್ರವೇಶಿಸಲು ನಾವು ಬಳಸುವ ಇಟರೇಟರ್‌ಗಳನ್ನು ರಚಿಸೋಣ. BERT ತನ್ನದೇ ಆದ ಎನ್‌ಕೋಡಿಂಗ್ ಫಂಕ್ಷನ್ ಅನ್ನು ಬಳಸುವುದರಿಂದ, ನಾವು ಮೊದಲು ನಿರ್ಧರಿಸಿದ `padify` ಗೆ ಸಮಾನವಾದ ಪ್ಯಾಡಿಂಗ್ ಫಂಕ್ಷನ್ ಅನ್ನು ವ್ಯಾಖ್ಯಾನಿಸಬೇಕಾಗುತ್ತದೆ:


```python
def pad_bert(b):
    # b is the list of tuples of length batch_size
    #   - first element of a tuple = label, 
    #   - second = feature (text sequence)
    # build vectorized sequence
    v = [tokenizer.encode(x[1]) for x in b]
    # compute max length of a sequence in this minibatch
    l = max(map(len,v))
    return ( # tuple of two tensors - labels and features
        torch.LongTensor([t[0] for t in b]),
        torch.stack([torch.nn.functional.pad(torch.tensor(t),(0,l-len(t)),mode='constant',value=0) for t in v])
    )

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=8, collate_fn=pad_bert, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=8, collate_fn=pad_bert)
```

ನಮ್ಮ ಪ್ರಕರಣದಲ್ಲಿ, ನಾವು `bert-base-uncased` ಎಂದು ಕರೆಯಲ್ಪಡುವ ಪೂರ್ವ-ಪ್ರಶಿಕ್ಷಿತ BERT ಮಾದರಿಯನ್ನು ಬಳಸಲಿದ್ದೇವೆ. `BertForSequenceClassfication` ಪ್ಯಾಕೇಜ್ ಬಳಸಿ ಮಾದರಿಯನ್ನು ಲೋಡ್ ಮಾಡೋಣ. ಇದು ನಮ್ಮ ಮಾದರಿಗೆ ವರ್ಗೀಕರಣಕ್ಕಾಗಿ ಅಗತ್ಯವಿರುವ ವಾಸ್ತುಶಿಲ್ಪವನ್ನು, ಅಂತಿಮ ವರ್ಗೀಕರಿಸುವುದನ್ನು ಸೇರಿಸಿ, ಈಗಾಗಲೇ ಹೊಂದಿರುವುದನ್ನು ಖಚಿತಪಡಿಸುತ್ತದೆ. ನೀವು ಅಂತಿಮ ವರ್ಗೀಕರಿಸುವುದರ ತೂಕಗಳು ಪ್ರಾರಂಭವಾಗಿಲ್ಲ ಎಂಬ ಎಚ್ಚರಿಕೆ ಸಂದೇಶವನ್ನು ಕಾಣುತ್ತೀರಿ, ಮತ್ತು ಮಾದರಿಗೆ ಪೂರ್ವ-ಪ್ರಶಿಕ್ಷಣ ಅಗತ್ಯವಿದೆ - ಅದು ಸಂಪೂರ್ಣವಾಗಿ ಸರಿಯಾಗಿದೆ, ಏಕೆಂದರೆ ಅದೇ ನಾವು ಮಾಡಬೇಕಾಗಿರುವುದು!


```python
model = transformers.BertForSequenceClassification.from_pretrained(bert_model,num_labels=4).to(device)
```

ಈಗ ನಾವು ತರಬೇತಿಯನ್ನು ಪ್ರಾರಂಭಿಸಲು ಸಿದ್ಧರಾಗಿದ್ದೇವೆ! BERT ಈಗಾಗಲೇ ಪೂರ್ವ-ತರಬೇತಿಗೊಳಿಸಲಾಗಿದೆ ಆದ್ದರಿಂದ ಪ್ರಾಥಮಿಕ ತೂಕಗಳನ್ನು ಹಾಳುಮಾಡದಂತೆ ಸಣ್ಣ ಕಲಿಕೆ ದರದಿಂದ ಪ್ರಾರಂಭಿಸಲು ನಾವು ಬಯಸುತ್ತೇವೆ.

ಎಲ್ಲಾ ಕಠಿಣ ಕೆಲಸವನ್ನು `BertForSequenceClassification` ಮಾದರಿ ಮಾಡುತ್ತದೆ. ತರಬೇತಿ ಡೇಟಾದ ಮೇಲೆ ಮಾದರಿಯನ್ನು ಕರೆಸಿದಾಗ, ಅದು ಇನ್‌ಪುಟ್ ಮಿನಿಬ್ಯಾಚ್‌ಗೆ ಲಾಸ್ ಮತ್ತು ನೆಟ್‌ವರ್ಕ್ ಔಟ್‌ಪುಟ್ ಎರಡನ್ನೂ ನೀಡುತ್ತದೆ. ನಾವು ಪ್ಯಾರಾಮೀಟರ್ ಆಪ್ಟಿಮೈಜೆಶನ್‌ಗೆ ಲಾಸ್ ಅನ್ನು ಬಳಸುತ್ತೇವೆ (`loss.backward()` ಬ್ಯಾಕ್ವರ್ಡ್ ಪಾಸ್ ಅನ್ನು ಮಾಡುತ್ತದೆ), ಮತ್ತು `out` ಅನ್ನು ತರಬೇತಿ ನಿಖರತೆಯನ್ನು ಲೆಕ್ಕಿಸಲು ಬಳಸುತ್ತೇವೆ, ಇದಕ್ಕಾಗಿ ಪಡೆದ ಲೇಬಲ್ಸ್ `labs` (`argmax` ಬಳಸಿ ಲೆಕ್ಕಿಸಲಾಗಿದೆ) ಅನ್ನು ನಿರೀಕ್ಷಿತ `labels` ಜೊತೆಗೆ ಹೋಲಿಸುತ್ತೇವೆ.

ಪ್ರಕ್ರಿಯೆಯನ್ನು ನಿಯಂತ್ರಿಸಲು, ನಾವು ಹಲವಾರು ಪುನರಾವೃತ್ತಿಗಳಲ್ಲಿ ಲಾಸ್ ಮತ್ತು ನಿಖರತೆಯನ್ನು ಸಂಗ್ರಹಿಸಿ, ಪ್ರತಿ `report_freq` ತರಬೇತಿ ಚಕ್ರಗಳಲ್ಲಿ ಅವುಗಳನ್ನು ಮುದ್ರಣ ಮಾಡುತ್ತೇವೆ.

ಈ ತರಬೇತಿ ಬಹಳ ಸಮಯ ತೆಗೆದುಕೊಳ್ಳಬಹುದು, ಆದ್ದರಿಂದ ನಾವು ಪುನರಾವೃತ್ತಿಗಳ ಸಂಖ್ಯೆಯನ್ನು ಮಿತಿಗೊಳಿಸುತ್ತೇವೆ.


```python
optimizer = torch.optim.Adam(model.parameters(), lr=2e-5)

report_freq = 50
iterations = 500 # make this larger to train for longer time!

model.train()

i,c = 0,0
acc_loss = 0
acc_acc = 0

for labels,texts in train_loader:
    labels = labels.to(device)-1 # get labels in the range 0-3         
    texts = texts.to(device)
    loss, out = model(texts, labels=labels)[:2]
    labs = out.argmax(dim=1)
    acc = torch.mean((labs==labels).type(torch.float32))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    acc_loss += loss
    acc_acc += acc
    i+=1
    c+=1
    if i%report_freq==0:
        print(f"Loss = {acc_loss.item()/c}, Accuracy = {acc_acc.item()/c}")
        c = 0
        acc_loss = 0
        acc_acc = 0
    iterations-=1
    if not iterations:
        break
```

ನೀವು ನೋಡಬಹುದು (ವಿಶೇಷವಾಗಿ ನೀವು ಪುನರಾವೃತ್ತಿಗಳ ಸಂಖ್ಯೆಯನ್ನು ಹೆಚ್ಚಿಸಿ ಮತ್ತು ಸಾಕಷ್ಟು ಸಮಯ ಕಾಯಿಸಿದರೆ) BERT ವರ್ಗೀಕರಣವು ನಮಗೆ ಉತ್ತಮ ನಿಖರತೆಯನ್ನು ನೀಡುತ್ತದೆ! ಇದಕ್ಕೆ ಕಾರಣ BERT ಈಗಾಗಲೇ ಭಾಷೆಯ ರಚನೆಯನ್ನು ಚೆನ್ನಾಗಿ ಅರ್ಥಮಾಡಿಕೊಂಡಿದೆ, ಮತ್ತು ನಾವು ಕೇವಲ ಅಂತಿಮ ವರ್ಗೀಕರಣೆಯನ್ನು ಸೂಕ್ಷ್ಮವಾಗಿ ಹೊಂದಿಸಬೇಕಾಗುತ್ತದೆ. ಆದರೆ, BERT ದೊಡ್ಡ ಮಾದರಿಯಾಗಿರುವುದರಿಂದ, ಸಂಪೂರ್ಣ ತರಬೇತಿ ಪ್ರಕ್ರಿಯೆಗೆ ಹೆಚ್ಚು ಸಮಯ ಬೇಕಾಗುತ್ತದೆ ಮತ್ತು ಗಂಭೀರ ಗಣನ ಶಕ್ತಿಯನ್ನು ಅಗತ್ಯವಿರುತ್ತದೆ! (GPU, ಮತ್ತು ಸಾಧ್ಯವಾದರೆ ಒಂದಕ್ಕಿಂತ ಹೆಚ್ಚು).

> **Note:** ನಮ್ಮ ಉದಾಹರಣೆಯಲ್ಲಿ, ನಾವು ಅತ್ಯಂತ ಸಣ್ಣ ಪೂರ್ವ-ತರಬೇತಿಗೊಂಡ BERT ಮಾದರಿಗಳಲ್ಲಿ ಒಂದನ್ನು ಬಳಸುತ್ತಿದ್ದೇವೆ. ಉತ್ತಮ ಫಲಿತಾಂಶಗಳನ್ನು ನೀಡುವ ಸಾಧ್ಯತೆ ಇರುವ ದೊಡ್ಡ ಮಾದರಿಗಳು ಇವೆ.


## ಮಾದರಿ ಕಾರ್ಯಕ್ಷಮತೆಯನ್ನು ಮೌಲ್ಯಮಾಪನ ಮಾಡುವುದು

ಈಗ ನಾವು ನಮ್ಮ ಮಾದರಿಯ ಕಾರ್ಯಕ್ಷಮತೆಯನ್ನು ಪರೀಕ್ಷಾ ಡೇಟಾಸೆಟ್ ಮೇಲೆ ಮೌಲ್ಯಮಾಪನ ಮಾಡಬಹುದು. ಮೌಲ್ಯಮಾಪನ ಲೂಪ್ ತರಬೇತಿ ಲೂಪಿನಂತೆ ಸಮಾನವಾಗಿದ್ದು, ಆದರೆ `model.eval()` ಅನ್ನು ಕರೆದು ಮಾದರಿಯನ್ನು ಮೌಲ್ಯಮಾಪನ ಮೋಡ್‌ಗೆ ಬದಲಾಯಿಸುವುದನ್ನು ಮರೆಯಬಾರದು.


```python
model.eval()
iterations = 100
acc = 0
i = 0
for labels,texts in test_loader:
    labels = labels.to(device)-1      
    texts = texts.to(device)
    _, out = model(texts, labels=labels)[:2]
    labs = out.argmax(dim=1)
    acc += torch.mean((labs==labels).type(torch.float32))
    i+=1
    if i>iterations: break
        
print(f"Final accuracy: {acc.item()/i}")
```

## ಸಾರಾಂಶ

ಈ ಘಟಕದಲ್ಲಿ, ನಾವು **transformers** ಗ್ರಂಥಾಲಯದಿಂದ ಪೂರ್ವ-ಪ್ರಶಿಕ್ಷಿತ ಭಾಷಾ ಮಾದರಿಯನ್ನು எளிதಾಗಿ ತೆಗೆದುಕೊಂಡು ನಮ್ಮ ಪಠ್ಯ ವರ್ಗೀಕರಣ ಕಾರ್ಯಕ್ಕೆ ಹೊಂದಿಕೊಳ್ಳಿಸುವುದನ್ನು ನೋಡಿದ್ದೇವೆ. ಅದೇ ರೀತಿಯಲ್ಲಿ, BERT ಮಾದರಿಗಳನ್ನು ಘಟಕ ಹೊರತೆಗೆಯುವಿಕೆ, ಪ್ರಶ್ನೆ ಉತ್ತರಿಸುವಿಕೆ ಮತ್ತು ಇತರ NLP ಕಾರ್ಯಗಳಿಗೆ ಬಳಸಬಹುದು.

Transformer ಮಾದರಿಗಳು NLP ನಲ್ಲಿ ಇತ್ತೀಚಿನ ಅತ್ಯುತ್ತಮ ತಂತ್ರಜ್ಞಾನವನ್ನು ಪ್ರತಿನಿಧಿಸುತ್ತವೆ, ಮತ್ತು ಬಹುತೇಕ ಸಂದರ್ಭಗಳಲ್ಲಿ ಕಸ್ಟಮ್ NLP ಪರಿಹಾರಗಳನ್ನು ಅನುಷ್ಠಾನಗೊಳಿಸುವಾಗ ಮೊದಲನೆಯದಾಗಿ ಪ್ರಯೋಗಿಸುವ ಪರಿಹಾರವಾಗಿರಬೇಕು. ಆದಾಗ್ಯೂ, ಈ ಘಟಕದಲ್ಲಿ ಚರ್ಚಿಸಲಾದ ಪುನರಾವರ್ತಿತ ನ್ಯೂರಲ್ ನೆಟ್‌ವರ್ಕ್‌ಗಳ ಮೂಲಭೂತ ತತ್ವಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವುದು, ನೀವು ಉನ್ನತ ಮಟ್ಟದ ನ್ಯೂರಲ್ ಮಾದರಿಗಳನ್ನು ನಿರ್ಮಿಸಲು ಬಯಸಿದರೆ ಅತ್ಯಂತ ಮುಖ್ಯವಾಗಿದೆ.
