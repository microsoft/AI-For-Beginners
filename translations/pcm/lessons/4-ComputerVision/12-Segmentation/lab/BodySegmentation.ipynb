For video production, like for weather forecast, we go need cut human image comot from camera and put am ontop another footage. Dis one dey usually happen wit **chroma key** techniques, wey dem go film human for front uniform color background, wey dem go later remove. For dis lab, we go train neural network model wey go fit cut human silhouette comot.

We go use [Segmentation Full Body MADS Dataset](https://www.kaggle.com/datasets/tapakah68/segmentation-full-body-mads-dataset) from Kaggle. Make you download di dataset manually from Kaggle and unzip am for di current directory.


```python
dataset_path = 'segmentation_full_body_mads_dataset_1192_img'

import os
import matplotlib.pyplot as plt
```

Make we see as di pictures for di dataset dey look like:


```python
img_path = os.path.join(dataset_path,'images')
mask_path = os.path.join(dataset_path,'masks')

fnames = os.listdir(img_path)

def load_image(img_name):
    img = plt.imread(os.path.join(img_path,img_name))
    mask = plt.imread(os.path.join(mask_path,img_name))
    return img,mask
```

```python
img, mask = load_image(fnames[5])

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(img)
ax[1].imshow(mask)
ax[0].axis('off')
ax[1].axis('off')
```
