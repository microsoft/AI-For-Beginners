## Assignment: Diophantine Equations

> Dis assignment na part of [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) and e dey inspired by [dis post](https://habr.com/post/128704/).

Your goal na to solve wetin dem dey call **Diophantine equation** - na equation wey get integer roots and integer coefficients. For example, make we look dis equation:

$$a+2b+3c+4d=30$$

You go need find integer roots $a$,$b$,$c$,$d\in\mathbb{N}$ wey go fit satisfy dis equation.

Hints:
1. You fit consider roots to dey inside di interval [0;30]
1. As gene, you fit use di list of root values
