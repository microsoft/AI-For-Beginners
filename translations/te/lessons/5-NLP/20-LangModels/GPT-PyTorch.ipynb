## OpenAI GPT తో ప్రయోగాలు

ఈ నోట్‌బుక్ [AI for Beginners Curriculum](http://aka.ms/ai-beginners) లో భాగం.

ఈ నోట్‌బుక్‌లో, Hugging Face `transformers` లైబ్రరీ ఉపయోగించి OpenAI-GPT మోడల్‌తో ఎలా ఆడుకోవచ్చో పరిశీలిస్తాము.

మరింత ఆలస్యం చేయకుండా, టెక్స్ట్ జనరేట్ చేసే పైప్‌లైన్‌ను సృష్టించి, జనరేషన్ ప్రారంభిద్దాం!


```python
from transformers import pipeline

model_name = 'openai-gpt' 

generator = pipeline('text-generation', model=model_name)

generator("Hello! I am a neural network, and I want to say that", max_length=100, num_return_sequences=5)
```

## ప్రాంప్ట్ ఇంజనీరింగ్

కొన్ని సమస్యల్లో, సరైన ప్రాంప్ట్‌లను రూపొందించడం ద్వారా మీరు వెంటనే openai-gpt జనరేషన్‌ను ఉపయోగించవచ్చు. క్రింది ఉదాహరణలను చూడండి:


```python
generator("Synonyms of a word cat:", max_length=20, num_return_sequences=5)
```

```python
generator("I love when you say this -> Positive\nI have myself -> Negative\nThis is awful for you to say this ->", max_length=40, num_return_sequences=5)
```

```python
generator("Translate English to French: cat => chat, dog => chien, student => ", top_k=50, max_length=30, num_return_sequences=3)
```

```python
generator("People who liked the movie The Matrix also liked ", max_length=40, num_return_sequences=5)
```

## టెక్స్ట్ శాంప్లింగ్ వ్యూహాలు

ఇప్పటివరకు మనం సాదారణ **గ్రీడి** శాంప్లింగ్ వ్యూహాన్ని ఉపయోగిస్తున్నాము, అంటే తదుపరి పదాన్ని అత్యధిక సంభావ్యత ఆధారంగా ఎంచుకుంటాము. ఇది ఎలా పనిచేస్తుందో ఇక్కడ ఉంది:


```python
prompt = "It was early evening when I can back from work. I usually work late, but this time it was an exception. When I entered a room, I saw"
generator(prompt,max_length=100,num_return_sequences=5)
```

**బీమ్ సెర్చ్** జనరేటర్‌కు పాఠ్య ఉత్పత్తి యొక్క అనేక దిశలను (*బీమ్స్*) అన్వేషించడానికి మరియు మొత్తం స్కోరు ఎక్కువగా ఉన్న వాటిని ఎంచుకోవడానికి అనుమతిస్తుంది. మీరు `num_beams` పారామీటర్‌ను అందించడం ద్వారా బీమ్ సెర్చ్ చేయవచ్చు. మీరు ఒక నిర్దిష్ట పరిమాణం యొక్క n-గ్రామ్స్ పునరావృతం అవ్వడాన్ని మోడల్‌కు శిక్షించడానికి `no_repeat_ngram_size` ను కూడా పేర్కొనవచ్చు:


```python
prompt = "It was early evening when I can back from work. I usually work late, but this time it was an exception. When I entered a room, I saw"
generator(prompt,max_length=100,num_return_sequences=5,num_beams=10,no_repeat_ngram_size=2)
```

**సాంప్లింగ్** మోడల్ ఇచ్చిన ప్రాబబిలిటీ పంపిణీని ఉపయోగించి తదుపరి పదాన్ని నిర్దిష్టంగా కాకుండా ఎంచుకుంటుంది. మీరు `do_sample=True` పారామీటర్ ఉపయోగించి సాంప్లింగ్‌ను ఆన్ చేయవచ్చు. మోడల్‌ను మరింత లేదా తక్కువ నిర్దిష్టంగా చేయడానికి మీరు `temperature` ను కూడా నిర్దేశించవచ్చు.


```python
prompt = "It was early evening when I can back from work. I usually work late, but this time it was an exception. When I entered a room, I saw"
generator(prompt,max_length=100,do_sample=True,temperature=0.8)
```

మేము శాంప్లింగ్‌కు అదనపు పారామితులను కూడా అందించవచ్చు:
* `top_k` శాంప్లింగ్ ఉపయోగించినప్పుడు పరిగణించాల్సిన పదాల ఎంపికల సంఖ్యను సూచిస్తుంది. ఇది మన టెక్స్ట్‌లో విచిత్రమైన (తక్కువ-సంభావ్యత) పదాలు రావడాన్ని తగ్గిస్తుంది.
* `top_p` కూడా ఇలానే ఉంటుంది, కానీ మేము మొత్తం సంభావ్యత p కంటే ఎక్కువ ఉన్న అత్యంత సంభావ్యమైన పదాల చిన్న ఉపసమూహాన్ని ఎంచుకుంటాము.

ఆ పారామితులను జోడించి ప్రయోగించడానికి స్వేచ్ఛగా ఉండండి.
