## అసైన్‌మెంట్: డయోఫాంటైన్ సమీకరణాలు

> ఈ అసైన్‌మెంట్ [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) లో భాగంగా ఉంది మరియు [ఈ పోస్ట్](https://habr.com/post/128704/) నుండి ప్రేరణ పొందింది.

మీ లక్ష్యం **డయోఫాంటైన్ సమీకరణం** అని పిలవబడే సమీకరణాన్ని పరిష్కరించడం - ఇది పూర్తి సంఖ్య మూలాలు మరియు పూర్తి సంఖ్య గుణకాలు కలిగిన సమీకరణ. ఉదాహరణకు, క్రింది సమీకరణాన్ని పరిశీలించండి:

$$a+2b+3c+4d=30$$

ఈ సమీకరణాన్ని సంతృప్తి పరచే పూర్తి సంఖ్య మూలాలు $a$,$b$,$c$,$d\in\mathbb{N}$ కనుగొనాలి.

సూచనలు:
1. మూలాలను [0;30] పరిధిలో ఉండేలా పరిగణించవచ్చు
1. ఒక జీన్‌గా, మూల విలువల జాబితాను ఉపయోగించవచ్చు
