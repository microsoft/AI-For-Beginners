# # ఎస్కేప్ కావడానికి మౌంటైన్ కార్‌ను శిక్షణ ఇవ్వడం

[AI for Beginners Curriculum](https://github.com/microsoft/ai-for-beginners) నుండి ల్యాబ్ అసైన్‌మెంట్.

మీ లక్ష్యం RL ఏజెంట్‌ను శిక్షణ ఇచ్చి OpenAI పరిసరంలో [Mountain Car](https://www.gymlibrary.ml/environments/classic_control/mountain_car/) ను నియంత్రించడం.

మనం పరిసరాన్ని సృష్టించడం ప్రారంభిద్దాం:


```python
import gym
env = gym.make('MountainCar-v0')
```

రాండమ్ ప్రయోగం ఎలా ఉంటుందో చూద్దాం:


```python
state = env.reset()
while True:
    env.render()
    action = env.action_space.sample()
    state, reward, done, info = env.step(action)
    if done:
        break
```
