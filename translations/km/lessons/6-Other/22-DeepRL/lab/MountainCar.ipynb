# # បណ្តុះបណ្តាល Mountain Car ឲ្យគេចចេញ

កិច្ចការពិសោធន៍ពី [កម្មវិធីសិក្សា AI សម្រាប់អ្នកចាប់ផ្តើម](https://github.com/microsoft/ai-for-beginners).

គោលបំណងរបស់អ្នកគឺបណ្តុះភ្នាក់ងារ RL ដើម្បីគ្រប់គ្រង [Mountain Car](https://www.gymlibrary.ml/environments/classic_control/mountain_car/) ក្នុងបរិស្ថាន OpenAI។

ចាប់ផ្តើមដោយបង្កើតបរិស្ថាន:


```python
import gym
env = gym.make('MountainCar-v0')
```

មកមើលថា ពិសោធន៍ចៃដន្យនេះមើលដូចម្តេច៖


```python
state = env.reset()
while True:
    env.render()
    action = env.action_space.sample()
    state, reward, done, info = env.step(action)
    if done:
        break
```

ឥឡូវនេះសៀវភៅកំណត់ត្រានេះជារបស់អ្នកទាំងអស់ - អ្នកអាចយកអាល់ហ្គរីធម៍ Policy Gradients និង Actor-Critic ពីមេរៀនមកអនុវត្តលើបញ្ហានេះបានដោយសេរី!
