## កិច្ចការ: សមីការឌីយូហ្វង់ទីន

> កិច្ចការនេះជាផ្នែកមួយនៃ [មេរៀន AI សម្រាប់អ្នកចាប់ផ្តើម](http://github.com/microsoft/ai-for-beginners) ហើយបានទទួលចbouរប្រភពពី [អត្ថបទនេះ](https://habr.com/post/128704/).

គោលបំណងរបស់អ្នកគឺដោះស្រាយដែលហៅថា **សមីការឌីយូហ្វង់ទីន** - សមីការមួយដែលមានដំណោះស្រាយជាចំនួនគត់ និងសមាសធាតុជាចំនួនគត់។ ឧទាហរណ៍ សូមពិចារណាសមីការដូចខាងក្រោម៖

$$a+2b+3c+4d=30$$

អ្នកត្រូវស្វែងរកដំណោះស្រាយជាចំនួនគត់ $a$,$b$,$c$,$d\in\mathbb{N}$ ដែលបំពេញសមីការនេះ។

Hints:
1. អ្នកអាចចាត់ទុកដំណោះស្រាយនៅក្នុងចន្លោះ [0;30]
1. ជាជីន សូមពិចារណាការប្រើបញ្ជីនៃតម្លៃដំណោះស្រាយ
