# бһҖбһ¶бһҡбһ’бҹ’бһңбһҫбһ…бҹҶбһҺбһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһӣбҹҒбһҒ MNIST бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ Framework бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһ•бҹ’бһ‘бһ¶бһӣбҹӢ

бһҖбһ¶бһҡбһ„бһ¶бһҡбһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһ–бһё [AI for Beginners Curriculum](https://github.com/microsoft/ai-for-beginners)бҹ”

### бһҖбһ¶бһҡбһўбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹҶбһҺбһ»бҹҶ

бһҖбһјбһҠбһ“бҹҒбҹҮбһ‘бһ¶бһүбһҷбһҖбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹҶбһҺбһ»бҹҶбһ–бһёбһғбҹ’бһӣбһ¶бҹҶбһ„бһҸбһ¶бһҳбһўбҹҠбһёбһ“бһ’бһәбһҺбһ·бһҸбҹ” бһўбҹ’бһ“бһҖбһҖбҹҸбһўбһ¶бһ…бһ…бһҳбҹ’бһӣбһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбҹҶбһҺбһ»бҹҶбһҠбҹ„бһҷбһҠбҹғбһ–бһёбһҗбһҸ `/data` бһҡбһ”бһҹбҹӢбһғбҹ’бһӣбһ¶бҹҶбһ„ AI Curriculum бһ•бһ„бһ”бһ¶бһ“бҹ”


```python
!rm *.pkl
!wget https://raw.githubusercontent.com/microsoft/AI-For-Beginners/main/data/mnist.pkl.gz
!gzip -d mnist.pkl.gz
```

```python
import pickle
with open('mnist.pkl','rb') as f:
    MNIST = pickle.load(f)
```

```python
labels = MNIST['Train']['Labels']
data = MNIST['Train']['Features']
```

бһҳбһҖбһҳбһҫбһӣбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһҷбһҫбһ„бһҳбһ¶бһ“бҹ–


```python
data.shape
```

### бһ”бҹҶбһ”бҹӮбһҖбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ

бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫ Scikit Learn бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹҶбһ”бҹӮбһҖбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҡбһңбһ¶бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ“бһ·бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бҹ–


```python
from sklearn.model_selection import train_test_split

features_train, features_test, labels_train, labels_test = train_test_split(data,labels,test_size=0.2)

print(f"Train samples: {len(features_train)}, test samples: {len(features_test)}")
```

### бһҹбҹҒбһ…бһҖбҹ’бһҸбһёбһҺбҹӮбһ“бһ¶бҹҶ

1. бһ…бһјбһҡбһ‘бһ¶бһүбһҷбһҖбһҖбһјбһҠбһ•бҹ’бһ‘бҹғбһ‘бҹ’бһҡбһ№бһҹбҹ’бһҸбһёбһ–бһёбһҳбҹҒбһҡбҹҖбһ“ бһ бһҫбһҷбһ”бһ·бһ‘бһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҹбҹҖбһңбһ—бҹ…бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ“бҹҒбҹҮ бһ¬ (бһӣбҹ’бһўбһҮбһ¶бһ„) бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҳбҹүбһјбһҢбһ»бһӣ Python бһ•бҹ’бһҹбҹҒбһ„
1. бһҖбҹҶбһҺбһҸбҹӢ бһ“бһ·бһ„вҖӢбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ perceptron бһҳбһҪбһҷбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ бһҠбҹ„бһҷбһҒбһ·бһҸбһҒбҹҶбһ¶бһҸбҹӢбһҸбһ¶бһҳбһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ“бһ·бһ„бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ—бһ¶бһ–бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһң бһҖбҹ’бһ“бһ»бһ„бһўбҹҶбһЎбһ»бһ„бһ–бҹҒбһӣбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ
1. бһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһҷбһӣбҹӢбһҗбһ¶ бһҸбһҫбһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһҳбһ¶бһ“бһҖбһ¶бһҡбһҸбһҳбҹ’бһҡбһ№бһҳбһ–бҹҒбһҖ (overfitting) бһ”бһҫбһҖбһҫбһҸбһЎбһҫбһ„ бһ¬бһўбһҸбҹӢ бһ бһҫбһҷбһҖбҹҶбһҺбһҸбҹӢбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһЎбһҫбһ„бһңбһ·бһү бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһ“бһ—бһ¶бһ–бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһң
1. бһҹбһ¶бһҡбһ¶бҹҶбһҹбһ¶бһҡбҹҒбһ“бһҮбҹҶбһ бһ¶бһ“бһҳбһ»бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ perceptron бҹў-бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ бһ“бһ·бһ„ бҹЈ-бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбҹ” бһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҮбһ¶бһҳбһҪбһҷбһҳбһ»бһҒбһ„бһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ‘бһёбһҡбһҪбһҳбһҡбһңбһ¶бһ„бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ
1. бһ–бҹ’бһҷбһ¶бһҷбһ¶бһҳбһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹҲ
    - бһҸбһҫбһҳбһ»бһҒбһ„бһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҡбһңбһ¶бһ„бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҳбһ¶бһ“бһҘбһҺбҹ’бһҢбһ·бһ–бһӣбһ‘бҹ…бһӣбһҫбһҖбһ¶бһҡбһ„бһ¶бһҡбһ”бһҺбҹ’бһҸбһ¶бһүбһ‘бҹҒ?
    - бһҸбһҫбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ¶бһү бҹў-бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ бһ¬ бҹЈ-бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹҒбһҹбһҖбһҖбһҳбҹ’бһҳбһ“бҹҒбҹҮбһ‘бҹҒ?
    - бһҸбһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҮбһҪбһ”бһ”бҹ’бһҡбһ‘бҹҲбһ”бһүбҹ’бһ бһ¶бһўбҹ’бһңбһёбһҒбҹ’бһӣбҹҮбһ“бҹ…бһ–бҹҒбһӣбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһ”бһҺбҹ’бһҸбһ¶бһү? бһҮбһ¶бһ–бһ·бһҹбҹҒбһҹбһ“бҹ…бһ–бҹҒбһӣбһ…бҹҶбһ“бһҪбһ“бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҖбһҫбһ“бһЎбһҫбһ„бҹ”
    - бһҸбһҫбһ‘бҹҶбһ„бһ“бҹӢбһҡбһ”бһҹбҹӢбһ”бһҺбҹ’бһҸбһ¶бһүбһ”бҹ’бһҡбһ–бҹ’бһҡбһ№бһҸбҹ’бһҸбһ·бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бһ“бҹ…бһ–бҹҒбһӣбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ? бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһ бһ¶бһүбһҖбҹ’бһҡбһ¶бһ”бһҸбһҳбҹ’бһӣбҹғбһҹбһҡбһ»бһ”бһўбһ¶бһ”бҹӢбһҹбһјбһҡбһ»бһҸбһ’бҹҶбһ”бҹҶбһ•бһ»бһҸбһ“бҹғбһ‘бҹҶбһ„бһ“бҹӢбһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһ“бһ№бһ„бһ…бҹҶбһ“бһҪбһ“бһҮбҹҶбһ“бһ¶бһ“бҹӢ бһҠбһҫбһҳбҹ’бһ”бһёбһҷбһӣбҹӢбһ–бһёбһ‘бҹҶбһ“бһ¶бһҖбҹӢбһ‘бҹҶбһ“бһ„бҹ”
