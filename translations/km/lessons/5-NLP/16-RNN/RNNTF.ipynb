# бһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһҷбһ“бҹ’бһҸбһңбһ·бһүбҹ’бһүбһ¶бһ”бһ“бһ”бһҸбҹ’бһҡ

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбҹүбһјбһҢбһ»бһӣбһҳбһ»бһ“ бһҷбһҫбһ„бһ”бһ¶бһ“бһӮбҹ’бһҡбһ”бһҠбһҺбҹ’бһҸбһ”бҹӢбһӣбһҫбһҖбһ¶бһҡбһҸбҹҶбһҺбһ¶бһ„бһ“бҹҗбһҷбһҹбһҳбҹ’бһ”бһјбһҡбһ”бҹӮбһ”бһ“бҹғбһўбһҸбҹ’бһҗбһ”бһ‘бҹ” бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ…бһ¶бһ”бҹӢбһҷбһҖбһ“бҹҗбһҷбһҹбһҡбһ»бһ”бһ“бҹғбһ–бһ¶бһҖбҹ’бһҷбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһҷбҹ„бһӮбһҳбһҪбһҷ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҳбһ·бһ“бһ”бһ¶бһ“бһӮбһ·бһҸбһ‘бһ»бһҖбһ“бһјбһң **бһӣбҹҶбһҠбһ¶бһ”бҹӢ** бһ“бҹғбһ–бһ¶бһҖбҹ’бһҷбһ‘бҹҒ бһ–бҹ’бһҡбҹ„бҹҮбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһ”бҹ’бһҡбһҳбһјбһӣбһ”бһүбҹ’бһ…бһјбһӣбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһўбһ“бһ»бһңбһҸбҹ’бһҸбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ”бһ¶бһ“бһҷбһҖбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ“бҹҒбҹҮбһ…бҹҒбһүбһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһҫбһҳбҹ” бһҠбҹ„бһҷбһҹбһ¶бһҡбһҗбһ¶бһҸбҹҶбһҺбһ¶бһ„бһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбһҳбһ·бһ“бһўбһ¶бһ…бһ”бһ„бҹ’бһ бһ¶бһүбһӣбҹҶбһҠбһ¶бһ”бҹӢбһ–бһ¶бһҖбҹ’бһҷбһ”бһ¶бһ“ бһ–бһҪбһҖбһңбһ¶бһҠбһјбһ…бһҮбһ¶бһҳбһ·бһ“бһўбһ¶бһ…бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһүбҹ’бһ бһ¶бһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү бһ¬бһ”бһүбҹ’бһ бһ¶бһҠбҹӮбһӣбһҳбһ¶бһ“бһ—бһ¶бһ–бһ…бҹ’бһҡбһ”бһјбһҖбһ…бҹ’бһҡбһ”бһӣбҹӢбһҠбһјбһ…бһҮбһ¶ бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһ¬бһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ¶бһ”бҹӢбһҷбһҖбһ“бҹҗбһҷбһ“бҹғбһҒбҹ’бһҹбҹӮбһўбһҖбҹ’бһҹбһҡбһҳбһҪбһҷ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ **бһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһҷбһ“бҹ’бһҸбһңбһ·бһүбҹ’бһүбһ¶бһҺбһ”бһҸбҹ’бһҡ**, бһ¬ RNNбҹ” бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ RNN бһӮбһә бһҷбһҫбһ„бһ•бҹ’бһүбһҫбһ”бҹ’бһҡбһҷбҹ„бһӮбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһІбҹ’бһҷбһ”бһҺбҹ’бһҸбһ¶бһүбһҳбһҪбһҷбһҠбҹ„бһҷбһҷбһҖбһҸбҹӮбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһҳбһҪбһҷбһ—бһ¶бһӮбһҸбһјбһ…бһҖбҹ’бһ“бһ»бһ„бһ–бҹҒбһӣбһҺбһ¶бһҳбһҪбһҷ бһ бһҫбһҷбһ”бһҺбҹ’бһҸбһ¶бһүбһ“бҹ„бҹҮбһ”бһүбҹ’бһ…бҹҒбһү **бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–** бһҳбһҪбһҷ бһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһ•бҹ’бһүбһҫбһҸбҹ’бһҡбһЎбһ”бҹӢбһ‘бҹ…бһ”бһҺбҹ’бһҸбһ¶бһүбһңбһ·бһүбһҮбһ¶бһҳбһҪбһҷбһ—бһ¶бһӮбһҸбһјбһ…бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ”

бһҠбҹ„бһҷбһўбһ“бһ»бһңбһҸбҹ’бһҸбһӣбҹҶбһҠбһ¶бһ”бҹӢбһ“бҹғбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷ $X_0,\dots,X_n$, RNN бһ”бһ„бҹ’бһҖбһҫбһҸбһҮбһ¶бһӣбҹҶбһҠбһ¶бһ”бҹӢбһ“бҹғбһ”бҹ’бһӣбһ»бһҖбһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһҷбһ“бҹ’бһҸбһңбһ·бһүбҹ’бһүбһ¶бһҺбһ”бһҸбҹ’бһҡ бһ бһҫбһҷбһ”бһ„бҹ’бһҡбҹҖбһ“бһӣбҹҶбһҠбһ¶бһ”бҹӢбһ“бҹҒбҹҮбһ–бһёбһҠбһҫбһҳбһҠбһӣбҹӢбһ…бһ”бҹӢбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ”бһҸбҹӢбһ”бһүбҹ’бһ…бҹ’бһҡбһ¶бһҹбһҸбһ¶бһҳбһңбһ·бһӣбһҖбҹ’бһҡбҹ„бһҷбҹ” бһ”бҹ’бһӣбһ»бһҖбһ”бһҺбҹ’бһҸбһ¶бһүбһ“бһёбһҳбһҪбһҷбҹ—бһ‘бһ‘бһҪбһӣбһӮбһј $(X_i,S_i)$ бһҮбһ¶бһҖбһ¶бһ“бҹӢбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣ бһ бһҫбһҷбһ”бһ„бҹ’бһ бһ¶бһү $S_{i+1}$ бһҮбһ¶бһӣбһ‘бҹ’бһ’бһ•бһӣбҹ” бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ $S_n$ бһ¬бһӣбһ‘бҹ’бһ’бһ•бһӣ $Y_n$ бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбһ¶бһҖбҹӢбһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһ…бһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһ”бҹӮбһ”бһўбһүбҹ’бһүбһҫбһүбһҡбһҖбһӣбһ‘бҹ’бһ’бһ•бһӣбҹ” бһ”бҹ’бһӣбһ»бһҖбһ”бһҺбҹ’бһҸбһ¶бһүбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһ…бҹӮбһҖбһҡбҹҶбһӣбҹӮбһҖбһ‘бһҳбҹ’бһ„бһ“бҹӢбһҠбһјбһ…бһӮбҹ’бһ“бһ¶ бһ бһҫбһҷбһ”бһ¶бһ“бһ”бһ„бҹ’бһҡбҹҖбһ“бһ–бһёбһҠбһҫбһҳбһҠбһӣбҹӢбһ…бһ”бҹӢбһҠбҹ„бһҷбһ”бһ„бҹ’бһңбһ·бһӣбһҸбҹӮбһҳбһҪбһҷбҹ”

> бһҡбһјбһ”бһ—бһ¶бһ–бһҒбһ¶бһ„бһӣбһҫбһ”бһ„бҹ’бһ бһ¶бһүбһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһҷбһ“бҹ’бһҸбһңбһ·бһүбҹ’бһүбһ¶бһҺбһ”бһҸбҹ’бһҡ бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ‘бһҳбҹ’бһҡбһ„бҹӢбһ”бһҳбҹүбҹ„бһ„бһ”бҹ’бһҡбҹӮбһҡбһ”бһ·бһ‘ (бһ“бҹ…бһҒбһ¶бһ„бһҶбҹ’бһңбҹҒбһ„) бһ“бһ·бһ„бһ‘бһҳбҹ’бһҡбһ„бҹӢбһҸбҹҶбһҺбһ¶бһ„бһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһ”бҹ’бһҡбһҳбһјбһӣ (бһ“бҹ…бһҒбһ¶бһ„бһҹбҹ’бһҠбһ¶бҹҶ)бҹ” бһңбһ¶бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҷбһӣбҹӢбһҗбһ¶ бһҖбҹ„бһҹбһ·бһҖбһ¶ RNN бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҳбһ¶бһ“ **бһ‘бһҳбҹ’бһ„бһ“бҹӢбһҠбҹӮбһӣбһўбһ¶бһ…бһ…бҹӮбһҖбһҡбҹҶбһӣбҹӮбһҖ** бһҠбһјбһ…бһӮбҹ’бһ“бһ¶бҹ”

бһҠбҹ„бһҷбһҹбһ¶бһҡvectorбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ– $S_0,\dots,S_n$ бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһҮбһјбһ“бһҸбһ¶бһҳбһ”бһҺбҹ’бһҸбһ¶бһү RNN бһўбһ¶бһ…бһҡбҹҖбһ“бһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’бһҸбһ¶бһҳбһӣбҹҶбһҠбһ¶бһ”бҹӢбһҡбһңбһ¶бһ„бһ–бһ¶бһҖбҹ’бһҷбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһҖбһ¶бһӣбһҺбһ¶бһ–бһ¶бһҖбҹ’бһҷ *not* бһ”бһ„бҹ’бһ бһ¶бһүбһ“бҹ…бһҒбһ¶бһ„бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһӣбҹҶбһҠбһ¶бһ”бҹӢ бһңбһ¶бһўбһ¶бһ…бһҡбҹҖбһ“бһ…бҹ’бһҡбһ¶бһ“бһ…бҹ„бһӣбһ’бһ¶бһҸбһ»бһҒбҹ’бһӣбҹҮбһ“бҹ…бһӣбһҫбһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бҹ”

бһҒбһ¶бһ„бһҖбҹ’бһ“бһ»бһ„ бһўбҹ’бһ“бһҖбһӣбҹҒбһ„бһҖбҹ„бһҹбһ·бһҖбһ¶ RNN бһҳбҹ’бһ“бһ¶бһҖбҹӢбһ“бһёбһҳбһҪбһҷбҹ—бһҳбһ¶бһ“бһҳбҹүбһ¶бһ‘бҹ’бһҡбһёбһҹбһ‘бһҳбҹ’бһ„бһ“бҹӢбһ–бһёбһҡ бһҠбһјбһ…бһҮбһ¶ $W_H$ бһ“бһ·бһ„ $W_I$ бһ“бһ·бһ„бһ•бҹ’бһ‘бһ»бһҖ bias $b$бҹ” бһ“бҹ…бһҮбҹҶбһ бһ¶бһ“ RNN бһ“бһёбһҳбһҪбһҷбҹ— бһҷбһҫбһ„бһҳбһ¶бһ“бһ”бһүбҹ’бһ…бһјбһӣ $X_i$ бһ“бһ·бһ„бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһ”бһүбҹ’бһ…бһјбһӣ $S_i$ бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһ…бҹҒбһүбһӣбҹ„бһҖбһӮбһ·бһҸбһҠбҹ„бһҷ $S_{i+1} = f(W_H\times S_i + W_I\times X_i+b)$ бһҠбҹӮбһӣ $f$ бһҮбһ¶бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһ”бҹҶбһӣбҹӮбһ„ (бһҮбһ¶бһ‘бһјбһ‘бҹ…бһҮбһ¶бһҡбһ¶бһ„ $\tanh$)бҹ”

> бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ бһ¶бһҠбһјбһ…бһҮбһ¶ бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ (бһҠбҹӮбһӣбһҷбһҫбһ„бһ“бһ№бһ„бһӮбҹ’бһҡбһ”бһҠбһҺбҹ’бһҸбһ”бҹӢбһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҸбҹ’бһҸбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢ) бһ¬бһ”бһҖбһ”бҹ’бһҡбҹӮбһҳбҹүбһ¶бһҹбҹҠбһёбһ“ бһҷбһҫбһ„бһҖбҹҸбһ…бһ„бҹӢбһ”бһ¶бһ“бһҸбһҳбҹ’бһӣбҹғбһӣбһ‘бҹ’бһ’бһ•бһӣбһ“бҹ…бһҡбҹҖбһ„бһҡбһ¶бһӣбҹӢбһҮбҹҶбһ бһ¶бһ“ RNN бһ•бһ„бһҠбҹӮбһҡбҹ” бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ бһҳбһ¶бһ“бһҳбҹүбһ¶бһ‘бҹ’бһҡбһёбһҹбһҳбһҪбһҷбһ‘бҹҖбһҸбһӮбһә $W_O$ бһ бһҫбһҷбһӣбһ‘бҹ’бһ’бһ•бһӣбһӮбһ·бһҸбһҗбһ¶ $Y_i=f(W_O\times S_i+b_O)$бҹ”

бһ…бһ„бҹӢбһІбҹ’бһҷбһҳбһҫбһӣбһҗбһ¶бһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҡбһҹбҹғбһ‘бһ“бҹ’бһ“бһ·бһ“бһҷбһ“бҹ’бһҸбһңбһ·бһүбҹ’бһүбһ¶бһҺбһ”бһҸбҹ’бһҡ бһўбһ¶бһ…бһҮбһҪбһҷбһҷбһҫбһ„бһ…бһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһ–бһёбһҖбһҳбҹ’бһҡбһ·бһҸбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһҗбҹ’бһҳбһёбһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”

> бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһҡбһ·бһңбҹҒбһҺ sandbox бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҡбһҸбҹӢбһҖбһјбһҠбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ бһҠбһҫбһҳбҹ’бһ”бһёбһ’бһ¶бһ“бһ¶бһҗбһ¶бһ…бҹӮбһҖбһ…бһ¶бһҷбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбҹҶбһЎбһҫбһ„ бһ“бһ·бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ‘бһ¶бһүбһҷбһҖбһҳбһ»бһ“бҹ” бһ”бһҫбһўбҹ’бһ“бһҖбһҡбһҸбҹӢбһ“бҹ…бһӣбһҫбһҖбһ»бҹҶбһ–бҹ’бһҷбһјбһ‘бҹҗбһҡбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҡбҹҶбһӣбһ„бһҖбһјбһҠбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ”

бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһё RNN бһҹбһ¶бһҳбһүбҹ’бһү, бһҳбһҪбһҷбһҜбһҖбһҸбһ¶бһҖбһ¶бһҡбһңбһ·бһӣбһҸбҹ’бһҡбһёбһҳбһ¶бһҸбҹ’бһҡбһӮбһәбһҮбһ¶бһ”бһҺбҹ’бһҸбһ¶бһүбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһҹбһ¶бһҳбһүбҹ’бһүбһҳбһҪбһҷ бһҠбҹӮбһӣбһ‘бһ‘бһҪбһӣбһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһҹбҹҶбһҺбһјбһӣ бһ“бһ·бһ„бһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһ”бҹҶбһҺбҹӮбһ„ бһ бһҫбһҷбһ•бһӣбһ·бһҸбһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһ”бҹҶбһҺбҹӮбһ„бһҗбҹ’бһҳбһёбҹ” бһҖбҹ’бһ“бһ»бһ„ Keras, бһ“бҹҒбҹҮбһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһ бһ¶бһүбһҠбҹ„бһҷбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ `SimpleRNN` бҹ”

бһ”бһҫбһ‘бҹ„бҹҮбһ”бһёбһ–бһҪбһҖбһҷбһҫбһ„бһўбһ¶бһ…бһ•бҹ’бһүбһҫбһҸбһјбһҖбҹӮбһ“бһҠбҹӮбһӣбһ”бһ¶бһ“бһҖбһјбһҠбһҮбһ¶ one-hot бһ‘бҹ…бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ RNN бһҸбҹ’бһҡбһ„бҹӢбҹ—бһҖбҹҸбһҠбҹ„бһҷ бһ“бҹ„бҹҮбһҳбһ·бһ“бһҳбҹӮбһ“бһҮбһ¶бһӮбҹҶбһ“бһ·бһҸбһӣбҹ’бһўбһ‘бҹҒ бһ–бҹ’бһҡбҹ„бҹҮбһҳбһ¶бһ“бһңбһ·бһҳбһ¶бһҸбҹ’бһҡбһҒбҹ’бһ–бһҹбҹӢбһ”бҹҶбһ•бһ»бһҸбҹ” бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ embedding бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ»бҹҮбһңбһ·бһҳбһ¶бһҸбҹ’бһҡбһҡбһ”бһҹбҹӢбһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһ–бһ¶бһҖбҹ’бһҷ бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ RNN бһ бһҫбһҷбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһҮбһ¶бһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһ…бҹҶбһҺбһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢ `Dense` бҹ”

> **бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ**: бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһҠбҹӮбһӣбһңбһ·бһҳбһ¶бһҸбҹ’бһҡбһҳбһ·бһ“бһҒбҹ’бһ–бһҹбҹӢбһ‘бҹҒ бһҮбһ¶бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ–бҹҒбһӣбһ”бҹ’бһҡбһҫ tokenization бһҖбһҳбҹ’бһҡбһ·бһҸбһўбһҖбҹ’бһҹбһҡ бһңбһ¶бһўбһ¶бһ…бһҳбһ¶бһ“бһ бҹҒбһҸбһ»бһ•бһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ•бҹ’бһүбһҫбһҸбһјбһҖбҹӮбһ“бһҠбҹӮбһӣбһ”бһ¶бһ“бһҖбһјбһҠбһҮбһ¶ one-hot бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбҹ„бһҹбһ·бһҖбһ¶ RNN бһ•бҹ’бһ‘бһ¶бһӣбҹӢбҹ”


```python
vocab_size = 20000

vectorizer = keras.layers.experimental.preprocessing.TextVectorization(
    max_tokens=vocab_size,
    input_shape=(1,))

model = keras.models.Sequential([
    vectorizer,
    keras.layers.Embedding(vocab_size, embed_size),
    keras.layers.SimpleRNN(16),
    keras.layers.Dense(4,activation='softmax')
])

model.summary()
```

> **бһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһҺбһ¶бҹҶбҹ–** бһҷбһҫбһ„бһ”бҹ’бһҡбһҫбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ embedding бһҳбһ·бһ“бһ”бһ¶бһ“бһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһ–бһҹбһ¶бһҳбһүбҹ’бһү бһ”бҹүбһ»бһ“бҹ’бһҸа№ҲаёІаёӘаёіаё«аёЈаёұаёҡбһӣбһ‘бҹ’бһ’бһ•бһӣбһӣбҹ’бһўбһҮбһ¶бһ„бһ“бҹҒбҹҮ бһҷбһҫбһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ embedding бһҠбҹӮбһӣбһ”бһ¶бһ“бһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһҡбһҪбһ…бһҮбһ¶бһҳбһҪбһҷ Word2Vec бһҠбһјбһ…бһ”бһ¶бһ“бһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһ“бҹ…бһҜбһҖбһҸбҹ’бһҸбһҮбҹҶбһ–бһјбһҖбһҳбһ»бһ“бҹ” бһңбһ¶бһҮбһ¶бһҖбһ¶бһҡбһ бһ¶бһҸбҹӢбһҠбҹҶбһҺбһҫбһҡбһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбһҳбҹ’бһҡбһјбһңбһҖбһјбһҠбһ“бҹҒбҹҮбһұбҹ’бһҷбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ embeddings бһҠбҹӮбһӣбһ”бһ¶бһ“бһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһҡбһҪбһ…бҹ”  

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҳбһҖбһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢ RNN бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ” RNN бһ‘бһјбһ‘бҹ…бһҳбһ¶бһ“бһ—бһ¶бһ–бһӣбҹҶбһ”бһ¶бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢ бһ–бҹ’бһҡбҹ„бҹҮбһ–бҹҒбһӣбһҠбҹӮбһӣбһҖбҹ„бһҹбһ·бһҖбһ¶бһҡбһ”бһҹбҹӢ RNN бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһҳбҹ’бһӣбҹӮбһ„ (unrolled) бһ‘бҹ…бһҸбһ¶бһҳбһ”бҹ’бһҡбһңбҹӮбһ„бһҡбһҠбҹ’бһӢбһҹбһ—бһ¶бһ– (sequence length) бһ…бҹҶбһ“бһҪбһ“бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҠбҹӮбһӣбһ…бһјбһӣбһҡбһҪбһҳбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҸбҹ’бһҡбһЎбһ”бҹӢбһҖбҹ’бһҡбҹ„бһҷ (backpropagation) бһӮбһәбһ“бһ№бһ„бһ’бҹҶбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бҹ” бһҠбһјбһ…бһ“бҹҒбҹҮбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһўбһҸбҹ’бһҡбһ¶бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶ (learning rate) бһҸбһјбһ…бһҮбһ¶бһ„бһҳбһ»бһ“ бһ бһҫбһҷбһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһ”бһҺбҹ’бһҸбһ¶бһүбһӣбһҫбһҢбһ¶бһҸбһ¶бһҹбһ·бһҖбҹ’бһҹбһ¶бһ’бҹҶбһҮбһ¶бһ„ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ¶бһ“бһӣбһ‘бҹ’бһ’бһ•бһӣбһӣбҹ’бһўбҹ” бһңбһ¶бһўбһ¶бһ…бһ…бһ¶бһӣбһҒбҹ’бһҹбҹӮбһ–бҹҒбһӣбһҷбһјбһҡбһҺбһ¶бһҹбҹӢ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫ GPU бһӮбһәбһ•бҹ’бһҸбһӣбҹӢбһўбһ¶бһ‘бһ·бһ—бһ¶бһ–бҹ”  

бһҠбһҫбһҳбҹ’бһ”бһёбһӣбҹ’бһ”бһҝбһ“бһЎбһҫбһ„ бһҷбһҫбһ„бһ“бһ№бһ„бһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһҳбҹүбһјбһҢбҹӮбһӣ RNN бһӣбһҫбһ…бҹҶбһҺбһ„бһҮбһҫбһ„бһҠбҹҶбһҺбһ№бһ„бһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮ бһҒбҹ’бһңбҹҮбһ–бһ¶бһҖбҹ’бһҷбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһҮбһ¶бһҳбһҪбһҷбһ–бһ¶бһҖбҹ’бһҷбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһҡбһҪбһҳ бһ“бһ·бһ„бһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһҳбҹүбһјбһҢбҹӮбһӣбһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһ”бһ¶бһ“бһ¬бһ‘бҹҒбҹ”


```python
def extract_title(x):
    return x['title']

def tupelize_title(x):
    return (extract_title(x),x['label'])

print('Training vectorizer')
vectorizer.adapt(ds_train.take(2000).map(extract_title))
```

> **бһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһҺбһ¶бҹҶ** бһҗбһ¶ бһ—бһ¶бһ–бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһ“бһ№бһ„бһ‘бһ¶бһ”бһҮбһ¶бһ„бһ“бҹҒбҹҮ бһҠбҹ„бһҷбһҹбһ¶бһҡбһҸбҹӮбһҷбһҫбһ„бһҖбҹҶбһ–бһ»бһ„бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһҸбҹӮбһӣбһҫбһ…бҹҶбһҺбһ„бһҮбһҫбһ„бһҠбҹҶбһҺбһ№бһ„бһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ”


## бһҖбһ¶бһҡбһҸбҹ’бһҡбһӣбһ”бҹӢбһҳбһҖбһңбһ·бһүбһӣбҹҶбһ бһ¶бһҸбҹӢбһҸбһҪбһўбһҖбҹ’бһҹбһҡбһ•бҹ’бһҹбҹҒбһ„бһ”бҹ’бһӣбҹӮбһҖ 

бһ…бһ„бһ…бһ¶бҹҶбһҗбһ¶бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ `TextVectorization` бһ“бһ№бһ„бһ”бһ“бҹ’бһҗбҹӮбһҳбһҹбһүбҹ’бһүбһ¶ pad бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹҶбһ–бҹҒбһүбһҒбҹ’бһҹбҹӮбһҸбһҪбһўбһҖбҹ’бһҹбһҡбһ•бҹ’бһҹбҹҒбһ„бһ”бҹ’бһҡбһңбҹӮбһ„бһҖбҹ’бһ“бһ»бһ„бһҳбһёбһ“бһёбһ”бһ¶бһ…бҹӢбһҠбҹ„бһҷбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·бҹ” бһңбһ¶бһҮбһ¶бһҖбҹӢбһҗбһ¶бһҹбһүбҹ’бһүбһ¶бһ‘бһ¶бҹҶбһ„бһ“бҹ„бҹҮбһҖбҹҸбһ…бһјбһӣбһҡбһҪбһҳбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһ•бһ„бһҠбҹӮбһҡ бһ бһҫбһҷбһңбһ¶бһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһҹбҹҶбһ”бҹүбһ¶бһҷбһҖбһ¶бһ“бһҸбһҳбҹ’бһҡбһјбһңбһ“бҹғбһҳбҹүбһјбһҠбҹӮбһӣбҹ”

бһҳбһ¶бһ“бһңбһ·бһ’бһёбһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһҷбһҫбһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһӣбҹүбһјбһҸбһ”бҹҶбһ•бһ»бһҸбһ”бһҡбһ·бһҳбһ¶бһҺбһҖбһ¶бһҡбһ”бҹҶбһ–бҹҒбһүбҹ” бһҖбҹ’бһ“бһ»бһ„бһ“бҹ„бҹҮбһҳбһ¶бһ“бһҮбҹҶбһ бһ¶бһ“бһҳбһҪбһҷбһӮбһәбһ’бҹ’бһңбһҫбһҡбҹҖбһ”бһ…бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҮбһ¶бһҹбҹҖбһӮбҹ’бһңбһёбһҸбһ¶бһҳбһ”бҹ’бһҡбһңбҹӮбһ„бһ“бҹғбһҒбҹ’бһҹбҹӮбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ бһ бһҫбһҷбһҖбҹ’бһҡбһ»бһҳбһӮбҹ’бһҡбһ”бҹӢбһҒбҹ’бһҹбҹӮбһҸбһҪбһўбһҖбҹ’бһҹбһҡбһҸбһ¶бһҳбһ‘бҹҶбһ бҹҶбҹ” бһңбһ¶бһўбһ¶бһ…бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ `tf.data.experimental.bucket_by_sequence_length` (бһҳбһҫбһӣ [бһҜбһҖбһҹбһ¶бһҡ](https://www.tensorflow.org/api_docs/python/tf/data/experimental/bucket_by_sequence_length))бҹ” 

бһңбһ·бһ’бһёбһҳбһҪбһҷбһ‘бҹҖбһҸбһӮбһәбһ”бҹ’бһҡбһҫ **masking**бҹ” бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ Keras бһҳбһ¶бһ“бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҒбҹ’бһӣбҹҮбһҠбҹӮбһӣбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҮбҹҶбһҡбһ»бһүбһ”бһүбҹ’бһ…бһјбһӣбһ”бһ“бҹ’бһҗбҹӮбһҳбһҠбҹӮбһӣбһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһҸбһҪбһўбһҖбҹ’бһҹбһҡбһҺбһ¶бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһӮбһ·бһҸбһ”бҹ’бһҡбһ¶бһҖбҹӢбһ‘бҹ…бһ–бҹҒбһӣбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбҹ” бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһјбһӣбһҖбһ¶бһҡбһҳбҹүбһ¶бһҹбҹ’бһҖбһёбһ„бһҖбҹ’бһ“бһ»бһ„бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһҷбһҫбһ„ бһҷбһҫбһ„бһўбһ¶бһ…бһ”бһүбҹ’бһ…бһјбһӣбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ `Masking` бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ ([бһҜбһҖбһҹбһ¶бһҡ](https://keras.io/api/layers/core_layers/masking/)) бһ¬бһ”бҹ’бһҡбһҫбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡ `mask_zero=True` бһ“бҹғбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ `Embedding` бһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ”

> **бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ**бҹ– бһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһ“бҹҒбҹҮбһ“бһ№бһ„бһ…бҹҶбһҺбһ¶бһҷбһ–бҹҒбһӣбһ”бҹ’бһҡбһ бҹӮбһӣ бҹҘ бһ“бһ¶бһ‘бһёбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбһҪбһҷ Epoch бһ“бҹ…бһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһҜбһҖбһҹбһ¶бһҡбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһҳбҹ’бһҡбһ¶бһҖбһ”бһ“бҹ’бһ‘бһ¶бһ“бҹӢбһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһ“бҹ…бһ–бҹҒбһӣбһҺбһ¶бһҖбҹҸбһ”бһ¶бһ“ бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһўбһҹбҹӢбһ…бҹҶбһҺбҹ„бһҳбҹ” бһўбҹ’бһ“бһҖбһҖбҹҸбһўбһ¶бһ…бһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһ“бһҪбһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷвҖӢбһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡвҖӢбһ”бһ“бҹ’бһҗбҹӮбһҳ `.take(...)` бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҢбһёбһҹбҹ’бһ‘бһәбһҸ `ds_train` бһ“бһ·бһ„ `ds_test`бҹ”


```python
def extract_text(x):
    return x['title']+' '+x['description']

def tupelize(x):
    return (extract_text(x),x['label'])

model = keras.models.Sequential([
    vectorizer,
    keras.layers.Embedding(vocab_size,embed_size,mask_zero=True),
    keras.layers.SimpleRNN(16),
    keras.layers.Dense(4,activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',metrics=['acc'], optimizer='adam')
model.fit(ds_train.map(tupelize).batch(batch_size),validation_data=ds_test.map(tupelize).batch(batch_size))
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһҖбҹҶбһ–бһ»бһ„бһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ”бҹҶбһӣбҹӮбһ„бһҳбһ»бһҒбһҳбһ¶бһҸбҹӢ (masking) бһ бһҫбһҷбһҷбһҫбһ„бһўбһ¶бһ…бһ”бһ„бҹ’бһҡбҹҖбһ“бһҳбҹүбһјбһҢбҹӮбһӣбһӣбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ”бһјбһҖбһ”бһүбҹ’бһ…бһјбһӣбһ‘бһ¶бҹҶбһ„бһҹбҹ’бһҡбһ»бһ„бһ“бҹғбһ…бҹҶбһҺбһ„бһҮбһҫбһ„ бһ“бһ·бһ„бһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бҹ”

> **бһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ**: бһҸбһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“вҖӢбһҠбһ№бһ„бһ‘бҹҒбһҗбһ¶бһҷбһҫбһ„бһ”бһ¶бһ“бһ”бҹ’бһҡбһҫ vectorizer бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҡбҹҖбһ“бһӣбһҫбһ…бҹҶбһҺбһ„бһҮбһҫбһ„бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“ бһ‘бҹҒ бһҳбһ·бһ“бһҳбҹӮбһ“бһӣбһҫбһҡбһ¶бһ„бһҖбһ¶бһҷбһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһ“бҹғбһўбһҸбҹ’бһҗбһ”бһ‘? бһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶ бһңбһ¶бһўбһ¶бһ…бһ”бһҺбҹ’бһҸбһ¶бһӣбһІбҹ’бһҷбһҒбҹ’бһӣбҹҮбһ“бҹғбһҸбһјбһҒбҹҒбһ“ бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҳбһҫбһӣбһҡбҹҶбһӣбһ„ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһӮбһҪбһҡбһҸбҹӮбһ”бһ„бҹ’бһҡбҹҖбһ“бһЎбһҫбһ„бһңбһ·бһү vectorizerбҹ” бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮ бһңбһ¶бһўбһ¶бһ…бһҳбһ¶бһ“бһ•бһӣбһ”бҹүбҹҮбһ–бһ¶бһӣбҹӢбһҸбһ·бһ…бһҸбһҪбһ…бһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһ“бҹ…бһҸбҹӮбһ”бҹ’бһҡбһҫ vectorizer бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҡбҹҖбһ“бһҳбһ»бһ“бһ“бҹҒбҹҮбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһ–бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбҹ”


## LSTM: бһўбһ„бҹ’бһӮбһ…бһ„бһ…бһ¶бҹҶбһҡбһҷбҹҲбһ–бҹҒбһӣбһҒбҹ’бһӣбһёбһ“бһ·бһ„бһҷбһјбһҡ

бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһ”бһүбҹ’бһ бһ¶бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ—вҖӢбһ“бҹғ RNN бһӮбһә **бһҖбҹҶбһҺбһҫбһ“ gradients бһҠбҹӮбһӣбһ”бһ¶бһҸбҹӢбһ”бһ„бҹӢ**бҹ” RNN бһўбһ¶бһ…бһҳбһ¶бһ“бһ”бҹ’бһҡбһңбҹӮбһ„бһҷбһјбһҡ бһ бһҫбһҷбһўбһ¶бһ…бһҳбһ¶бһ“бһҖбһ¶бһҡбһӣбҹҶбһ”бһ¶бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһјбһ“ gradients бһҸбҹ’бһҡбһЎбһ”бҹӢбһ‘бҹ…бһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢбһ‘бһёбһҳбһҪбһҷбһ“бҹғбһ”бһҺбҹ’бһҸбһ¶бһүбһ“бҹ…бһ–бҹҒбһӣ backpropagationбҹ” бһ“бҹ…бһ–бҹҒбһӣбһ“бҹҒбҹҮбһҖбһҫбһҸбһЎбһҫбһ„ бһ“бҹҒбҹҮбһ”бһҺбҹ’бһҸбһ¶бһүбһҳбһ·бһ“бһўбһ¶бһ…бһҡбҹҖбһ“бһ‘бҹҶбһ“бһ¶бһҖбҹӢбһ‘бҹҶбһ“бһ„бһҡбһңбһ¶бһ„ token бһҠбҹӮбһӣбһҹбҹ’бһҗбһ·бһҸбһҶбҹ’бһ„бһ¶бһҷбһӮбҹ’бһ“бһ¶бһ”бһ¶бһ“бһ‘бҹҒбҹ” бһҳбһјбһӣбһ бҹҒбһҸбһ»вҖӢбһҳбһҪбһҷбһҠбһҫбһҳбҹ’бһ”бһёбһҮбҹҖбһҹбһңбһ¶бһ„бһ”бһүбҹ’бһ бһ¶бһ“бҹҒбҹҮбһӮбһә бһҠбһ¶бһҖбҹӢбһ”бһ“бҹ’бһҗбҹӮбһҳ **бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҷбҹүбһ¶бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢ** бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ **бһ‘бҹ’бһңбһ¶бһҡбһҮбҹ’бһҡбһ¶бһ”**бҹ” бһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһҠбҹӮбһӣбһҮбһ¶бһ‘бһјбһ‘бҹ…бһ”бһ„бҹ’бһ бһ¶бһүбһ‘бҹ’бһңбһ¶бһҡбһҮбҹ’бһҡбһ¶бһ”бһӮбһә **бһўбһ„бҹ’бһӮбһ…бһ„бһ…бһ¶бҹҶбһҡбһҷбҹҲбһ–бҹҒбһӣбһҒбҹ’бһӣбһёбһ“бһ·бһ„бһҷбһјбһҡ** (LSTM) бһ“бһ·бһ„ **бһҜбһҖбһҸбһ¶бһҮбҹ’бһҡбһ¶бһ”бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ•бҹ’бһҹбһ–бҹ’бһңбһ•бҹ’бһҹбһ¶бһҷ** (GRU)бҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһўбҹҶбһ–бһё LSTM бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбҹ”

бһ”бһҺбҹ’бһҸбһ¶бһү LSTM бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҡбҹҖбһ”бһ…бҹҶбһҮбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҠбһјбһ… RNN бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҳбһ¶бһ“бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһ–бһёбһҡбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһҮбһјбһ“бһ–бһёбһҮбһ¶бһ“бҹӢбһҳбһҪбһҷбһ‘бҹ…бһҮбһ¶бһ“бҹӢбһҳбһҪбһҷ бһӮбһәбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһ–бһ·бһҸ $c$ бһ“бһ·бһ„ бһңбһ·бһ‘бҹҗбһҡбһӣбһ¶бһҖбҹӢ $h$бҹ” бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҸбһ¶бһ‘бһёбһҳбһҪбһҷ бһңбһ·бһ‘бҹҗбһҡбһӣбһ¶бһҖбҹӢ $h_{t-1}$ бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„бһңбҹүбһ·бһ…бһ‘бҹҗбһҡ input $x_t$ бһ бһҫбһҷбһ–бһҪбһҖбһңбһ¶бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһўбҹ’бһңбһёбһҠбҹӮбһӣбһҖбһҫбһҸбһЎбһҫбһ„бһ…бҹҶбһ–бҹ„бҹҮбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ– $c_t$ бһ“бһ·бһ„бһ…бҹҒбһү $h_{t}$ бһҸбһ¶бһҳбһҡбһҷбҹҲ **бһ‘бҹ’бһңбһ¶бһҡбһҮбҹ’бһҡбһ¶бһ”**бҹ” бһ‘бҹ’бһңбһ¶бһҡвҖӢбһҡбҹҖбһ„бһҡбһ¶бһӣбҹӢбһ‘бҹ’бһңбһ¶бһҡбһҳбһ¶бһ“бһӮбһҳбҹ’бһӣбһ¶бһҸ sigmoid (бһ…бҹҒбһүбһҖбҹ’бһ“бһ»бһ„бһҮбһҪбһҡбһ“бҹғ $[0,1]$) бһҠбҹӮбһӣбһўбһ¶бһ…бһӮбһ·бһҸбһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖ bitwise mask бһ“бҹ…бһ–бҹҒбһӣбһӮбһ»бһҺбһҮбһ¶бһҳбһҪбһҷбһңбһ·бһ‘бҹҗбһҡбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бҹ” LSTM бһҳбһ¶бһ“бһ‘бҹ’бһңбһ¶бһҡбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ (бһ–бһёбһҶбҹ’бһңбҹҒбһ„бһ‘бҹ…бһҹбҹ’бһҸбһ¶бҹҶбһ“бҹ…бһӣбһҫбһҡбһјбһ”бһ—бһ¶бһ–бһҒбһ¶бһ„бһӣбһҫ):
* **бһ‘бҹ’бһңбһ¶бһҡбһ—бҹ’бһӣбҹҒбһ…бһ”бһ¶бһҸбҹӢ** бһҠбҹӮбһӣбһҖбҹҶбһҺбһҸбҹӢбһҗбһ¶ бһ’бһ¶бһҸбһ»бһҺбһ¶бһҠбҹӮбһӣбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһ·бһ‘бҹҗбһҡ $c_{t-1}$ бһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ—бҹ’бһӣбҹҒбһ…бһ”бһ¶бһҸбҹӢ бһ бһҫбһҷбһ’бһ¶бһҸбһ»бһҺбһ¶бһҠбҹӮбһӣбһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһҮбһјбһ“бһҸбһ‘бҹ…бҹ”
* **бһ‘бҹ’бһңбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣ** бһҠбҹӮбһӣбһҖбҹҶбһҺбһҸбҹӢбһҗбһ¶бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ”бҹүбһ»бһ“бҹ’бһҳбһ¶бһ“бһ–бһёбһңбһ·бһ‘бҹҗбһҡбһ”бһүбҹ’бһ…бһјбһӣбһ“бһ·бһ„бһңбһ·бһ‘бҹҗбһҡбһӣбһ¶бһҖбҹӢбһҳбһ»бһ“бһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһ·бһ‘бҹҗбһҡбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бҹ”
* **бһ‘бҹ’бһңбһ¶бһҡбһ…бҹҒбһү** бһҠбҹӮбһӣбһҷбһҖбһңбһ·бһ‘бҹҗбһҡбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҗбҹ’бһҳбһё бһ бһҫбһҷбһҹбһҳбҹ’бһҡбҹҒбһ…бһҗбһ¶бһ’бһ¶бһҸбһ»бһҺбһ¶бһҠбҹӮбһӣ бһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһңбһ·бһ‘бҹҗбһҡбһӣбһ¶бһҖбҹӢбһҗбҹ’бһҳбһё $h_t$бҹ”

бһ’бһ¶бһҸбһ»бһ“бҹғбһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ– $c$ бһўбһ¶бһ…бһӮбһ·бһҸбһҮбһ¶бһ‘бһ„бҹӢбһҠбҹӮбһӣбһўбһ¶бһ…бһ”бһ·бһ‘ бһ“бһ·бһ„бһ”бһҫбһҖбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһ“бҹ…бһ–бҹҒбһӣбһҷбһҫбһ„бһҮбһҪбһ”бһҲбҹ’бһҳбҹ„бҹҮ *Alice* бһҖбҹ’бһ“бһ»бһ„бһӣбҹҶбһҠбһ¶бһ”бҹӢбһ–бһ¶бһҖбҹ’бһҷ бһҷбһҫбһ„бһӮбһ·бһҸбһҗбһ¶бһҲбҹ’бһҳбҹ„бҹҮбһ“бҹҒбҹҮбһ бҹ…бһҠбһӣбҹӢбһҹбҹ’бһҸбҹ’бһҡбһё бһ бһҫбһҷбһ”бһҫбһҖбһ‘бһ„бҹӢбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһҷбһҫбһ„бһҳбһ¶бһ“бһ“бһ¶бһҳбһҹбҹ’бһҡбһёбһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһҷбҹ„бһӮбҹ” бһ“бҹ…бһ–бҹҒбһӣбһҷбһҫбһ„бһ”бһ“бҹ’бһҸбһҮбһҪбһ”бһ–бһ¶бһҖбҹ’бһҷ *and Tom* бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһҫбһҖбһ‘бһ„бҹӢбһҠбҹӮбһӣбһ”бһ„бҹ’бһ бһ¶бһүбһҗбһ¶бһҷбһҫбһ„бһҳбһ¶бһ“бһ“бһ¶бһҳбһ–бһ бһ»бһ”бһ»бһӮбҹ’бһӮбһӣбҹ” бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҠбҹ„бһҷбһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ– бһҷбһҫбһ„бһўбһ¶бһ…бһҸбһ¶бһҳбһҠбһ¶бһ“бһӣбһҖбҹ’бһҒбһҺбҹҲбһңбҹҒбһҷбҹ’бһҷбһ¶бһҖбһҡбһҺбҹҚбһ“бҹғбһ”бҹ’бһҡбһҷбҹ„бһӮбҹ”

> **бһҖбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ**: бһ“бҹҒбҹҮбһҮбһ¶бһ—бһҹбҹ’бһҸбһ»бһҸбһ¶бһ„бһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҷбһӣбҹӢбһ–бһёбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҒбһ¶бһ„бһҖбҹ’бһ“бһ»бһ„бһ“бҹғ LSTM: [Understanding LSTM Networks](https://colah.github.io/posts/2015-08-Understanding-LSTMs/) бһҠбҹ„бһҷ Christopher Olahбҹ”

бһ“бҹ…бһ–бҹҒбһӣбһ”бһ“бҹ’бһ‘бҹҮбһҒбһ¶бһ„бһҖбҹ’бһ“бһ»бһ„бһ“бҹғбһҖбҹ„бһҹбһ·бһҖбһ¶ LSTM бһўбһ¶бһ…бһҳбһҫбһӣбһ‘бҹ…бһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү бһҖбһ¶бһҡбһ”бҹ’бһҡбҹӮбһ”бҹ’бһҡбһҪбһӣбһ“бҹҒбҹҮбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһӣбһ¶бһҖбҹӢбһҒбһ¶бһ„бһҖбҹ’бһ“бһ»бһ„бһҮбһ¶бһ“бҹӢ `LSTM` бһҡбһ”бһҹбҹӢ Keras бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһўбҹ’бһңбһёбһҠбҹӮбһӣбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ’бҹ’бһңбһҫбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҒбһ¶бһ„бһӣбһҫбһӮбһәбһ”бҹ’бһҠбһјбһҡбһҮбһ¶бһ“бҹӢ recurrent бҹ”


```python
model = keras.models.Sequential([
    vectorizer,
    keras.layers.Embedding(vocab_size, embed_size),
    keras.layers.LSTM(8),
    keras.layers.Dense(4,activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',metrics=['acc'], optimizer='adam')
model.fit(ds_train.map(tupelize).batch(8),validation_data=ds_test.map(tupelize).batch(8))
```

> **бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ** бһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ LSTM бһ•бһ„бһҠбҹӮбһҡ бһҳбһ¶бһ“бһӣбҹ’бһ”бһҝбһ“бһҷбһәбһҸ бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҳбһ·бһ“бһғбһҫбһүбһҖбһ¶бһҡбһҖбһҫбһ“бһЎбһҫбһ„бһҒбҹ’бһӣбҹҮбҹ—бһ“бҹғбһ—бһ¶бһ–бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһң бһ“бҹ…бһҠбһҫбһҳбһ“бҹғбһҖбһ¶бһҡбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһ“бҹ„бҹҮбһ‘бҹҒбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ“бҹ’бһҸбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһҡбһҷбҹҲбһ–бҹҒбһӣбһҳбһҪбһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ—бһ¶бһ–бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбһӣбҹ’бһўбҹ”


## RNN бһ‘бһ·бһҹвҖӢбһ”бһјбһҖ бһ“бһ·бһ„ RNN бһ”бһ…бҹ’бһҡбһ”бһ¶бҹҶбһ„вҖӢбһ…бҹ’бһҡбһҫбһ“вҖӢбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢ

бһҖбҹ’бһ“бһ»бһ„бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһ‘бҹ…бһҡбһҪбһ…бһ“бҹҒбҹҮ бһ”бһҺбҹ’бһҠбһ¶бһү recurrent бһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһ–бһёбһҠбһҫбһҳбһҮбһҪбһҡвҖӢбһҠбһӣбҹӢбһ…бһ»бһ„бһҮбһҪбһҡбҹ” бһңбһ¶бһ бһ¶бһҖбҹӢбһҠбһјбһ…бһҮбһ¶вҖӢбһ’бһҳбҹ’бһҳбһҸбһ¶бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҷбһҫбһ„ бһ–бһёбһ–бҹ’бһҡбҹ„бҹҮбһңбһ¶бһҹбҹ’бһҡбһ”бһ‘бҹ…бһҸбһ¶бһҳбһ‘бһ·бһҹбһҠбҹ…бһҠбһҠбҹӮбһӣбһ“бҹғбһҖбһ¶бһҡбһўбһ¶бһ“бһ¬бһҖбһ¶бһҡбһҹбҹ’бһҸбһ¶бһ”бҹӢбһҹбҹҶбһЎбҹҒбһ„бҹ” бһ‘бҹ„бҹҮбһҮбһ¶бһҷбҹүбһ¶бһ„бһҺбһ¶ бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ…бһјбһӣбһ”бҹ’бһҡбһҫбһҮбһҪбһҡбһҠбһ¶бһҖбҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹ„бһҷбһ…бҹғбһҠбһ“бҹ’бһҷ бһңбһ¶бһҳбһ¶бһ“бһ бҹҒбһҸбһ»бһ•бһӣбһҗбһ¶ бһӮбһҪбһҡбһұбҹ’бһҷбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҮбһ¶бһ“бҹӢбһ”бһҺбҹ’бһҠбһҪбһҸбһ·бһ„бһ‘бһ·бһҹбһ–бһёбһҡбһ‘бһ·бһҹбҹ” RNN бһҠбҹӮбһӣбһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһӮбһҺбһ“бһ¶бһҺбҹҚбһ–бһёбһ‘бһ·бһҹбһ–бһёбһҡ бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ…бһҗбһ¶ **bidirectional** RNN бһ бһҫбһҷбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһ¶бһ”бһ¶бһ“бһҠбҹ„бһҷбһ”бһүбҹ’бһҲбһҡбһҮбһ¶бһ“бҹӢ recurrent бһҮбһ¶бһҳбһҪбһҷбһҮбһ¶бһ“бҹӢбһ–бһ·бһҹбҹҒбһҹ `Bidirectonal`бҹ”

> **бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ**бҹ– бһҮбһ¶бһ“бҹӢ `Bidirectional` бһ”бһ„бҹ’бһҖбһҫбһҸбһ…бһҳбҹ’бһӣбһ„бһ–бһёбһҡбһ“бҹғбһҮбһ¶бһ“бҹӢбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһ¶ бһ бһҫбһҷбһҖбҹҶбһҺбһҸбҹӢбһӣбһҖбҹ’бһҒбһҺбҹҲ `go_backwards` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһҳбҹ’бһӣбһ„бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ‘бҹ…бһҮбһ¶ `True` бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһ‘бһ·бһҹбһңбһ·бһүбһ“бҹ…бһҳбһ»бһҒбһҮбһҪбһҡбһҠбһ¶бһҖбҹӢбҹ” 

бһ”бһҺбҹ’бһҠбһ¶бһү recurrent бһ‘бһ¶бҹҶбһ„бһҮбһ¶бһ“бҹӢбһҸбҹӮбһҳбһҪбһҷ (unidirectional) бһ¬бһ‘бҹ’бһңбһ·бһ‘бһ·бһҹ (bidirectional) бһ–бһ“бҹ’бһӣбһҝбһ“бһҸбһҳбҹ’бһҡбһ„бһҖбҹ’бһ“бһ»бһ„бһҳбһҪбһҷбһҮбһҪбһҡ бһ бһҫбһҷбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһңбһ¶бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбҹүбһ·бһҖбһ‘бҹҗбһҡбһҡбһҠбҹ’бһӢ бһ¬бһҸбҹ’бһҡбһЎбһ”бҹӢбһңбһ¶бһҮбһ¶бһ•бһӣбһ…бҹҒбһүбҹ” бһҠбһјбһ…бһҮбһ¶ бһ”бһҺбҹ’бһҠбһ¶бһүбһҳбһ¶бһ“бһҖбһ¶бһҡбһҹбҹ’бһ„бҹ„ (convolutional networks) бһ•бһ„бһҠбҹӮбһҡ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҮбһ¶бһ“бҹӢ recurrent бһҗбҹ’бһҳбһёбһҳбһҪбһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҮбһ¶бһ“бҹӢбһҠбҹҶбһ”бһјбһ„ бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ¶бһ”бҹӢбһҷбһҖбһ•бҹ’бһ‘бһ¶бҹҶбһ„бһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢ бһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһЎбһҫбһ„бһ–бһёбһ•бҹ’бһ‘бһ¶бҹҶбһ„бһҖбһҳбҹ’бһҡбһ·бһҸбһ‘бһ¶бһ”бһҠбҹӮбһӣбһ…бһ¶бһ”бҹӢбһҷбһҖбһҠбҹ„бһҷбһҮбһ¶бһ“бҹӢбһҠбҹҶбһ”бһјбһ„бҹ” бһңбһ¶бһҠбһ№бһҖбһ“бһ¶бҹҶбһҷбһҫбһ„бһҠбһӣбҹӢбһӮбҹҶбһ“бһ·бһҸбһ“бҹғ **multi-layer RNN** бһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бһҺбҹ’бһҠбһ¶бһ—бһ¶бһ–бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ“бҹӢ recurrent бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһ–бһёбһҡбһҖбһ“бҹ’бһӣбҹӮбһ„бһҠбҹӮбһӣбһ•бһӣбһ…бҹҒбһүбһ–бһёбһҮбһ¶бһ“бҹӢбһҳбһ»бһ“бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ•бҹ’бһүбһҫбһ‘бҹ…бһҮбһ¶бһҮбһҪбһҡбһҠбһ¶бһҖбҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҡбһ”бһҹбҹӢбһҮбһ¶бһ“бҹӢбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ”

бһ…бҹҶбһҺбһ¶бҹҶбһҮбһјбһ“ бһҷбһҫбһ„бһ“бһ№бһ„бһҹбһ„бҹӢ LSTM бһ‘бҹ’бһңбһ·бһ‘бһ·бһҹбһ–бһёбһҡбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ бһ¶бһҖбһ¶бһҡбһ”бҹӮбһ„бһ…бҹӮбһҖбһ…бҹҶбһҺбһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ”

> **бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢ** бҹ– бһҖбһјбһҠбһ“бҹҒбҹҮбһҳбҹ’бһҸбһ„бһ‘бҹҖбһҸбһ…бҹҶбһҺбһ¶бһҷбһ–бҹҒбһӣбһңбҹҒбһӣбһ¶бһҒбҹ’бһӣбҹҮ бһҸбҹӮбһңбһ¶бһ•бҹ’бһҸбһӣбҹӢбһ“бһјбһңбһ—бһ¶бһ–бһҸбҹ’бһҡбһәбһҳбһҸбҹ’бһҡбһјбһңбһҒбҹ’бһ–бһҹбҹӢбһ”бҹҶбһ•бһ»бһҸ бһҠбҹӮбһӣбһҷбһҫбһ„бһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһҠбһӣбҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮбҹ” бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһӮбһҪбһҡбһӣбһҫбһҖбһ‘бһ№бһҖбһ…бһ·бһҸбҹ’бһҸбһҡбһ„бҹӢбһ…бһ¶бҹҶ бһ“бһ·бһ„бһҳбһҫбһӣбһӣбһ‘бҹ’бһ’бһ•бһӣбҹ”


```python
model = keras.models.Sequential([
    vectorizer,
    keras.layers.Embedding(vocab_size, 128, mask_zero=True),
    keras.layers.Bidirectional(keras.layers.LSTM(64,return_sequences=True)),
    keras.layers.Bidirectional(keras.layers.LSTM(64)),    
    keras.layers.Dense(4,activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',metrics=['acc'], optimizer='adam')
model.fit(ds_train.map(tupelize).batch(batch_size),
          validation_data=ds_test.map(tupelize).batch(batch_size))
```

## RNN бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ„бһ¶бһҡбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ

бһҡбһ бһјбһҸбһҳбһҖбһҠбһӣбҹӢбһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ”бһ¶бһ“бһ•бҹ’бһҸбҹ„бһҸбһӣбһҫбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ RNN бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ¶бһҸбҹӢбһҗбҹ’бһ“бһ¶бһҖбҹӢбһӣбҹҶбһҠбһ¶бһ”бҹӢбһўбһҖбҹ’бһҹбһҡбһ”бҹ’бһҡбһҷбҹ„бһӮбҹ” бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһңбһ¶бһўбһ¶бһ…бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҖбһ¶бһҡбһ„бһ¶бһҡбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ”бһ¶бһ“бһ…бҹ’бһҡбһҫбһ“ бһҠбһјбһ…бһҮбһ¶бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһ“бһ·бһ„бһҖбһ¶бһҡбһ”бҹ’бһҡбҹӮбһ—бһ¶бһҹбһ¶бһӮбһҺбһ“бһ¶ &mdash; бһҷбһҫбһ„бһ“бһ№бһ„бһ–бһ·бһ…бһ¶бһҡбһҺбһ¶бһҖбһ¶бһҡбһ„бһ¶бһҡбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбһ“бҹ…бһҜбһҖбһ·бһҸбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ”
