## അസൈൻമെന്റ്: ഡയോഫന്റൈൻ സമവാക്യങ്ങൾ

> ഈ അസൈൻമെന്റ് [AI for Beginners Curriculum](http://github.com/microsoft/ai-for-beginners) എന്ന കോഴ്സിന്റെ ഭാഗമാണ്, കൂടാതെ [ഈ പോസ്റ്റ്](https://habr.com/post/128704/) നിന്നാണ് പ്രചോദനം ലഭിച്ചത്.

നിങ്ങളുടെ ലക്ഷ്യം **ഡയോഫന്റൈൻ സമവാക്യം** എന്നറിയപ്പെടുന്ന സമവാക്യം പരിഹരിക്കുക എന്നതാണ് - ഇത് പൂർണ്ണസംഖ്യ മൂല്യങ്ങളും പൂർണ്ണസംഖ്യ ഗുണകങ്ങളും ഉള്ള സമവാക്യമാണ്. ഉദാഹരണത്തിന്, താഴെ കാണുന്ന സമവാക്യം പരിഗണിക്കാം:

$$a+2b+3c+4d=30$$

ഈ സമവാക്യം തൃപ്തിപ്പെടുത്തുന്ന പൂർണ്ണസംഖ്യ മൂല്യങ്ങൾ $a$,$b$,$c$,$d\in\mathbb{N}$ കണ്ടെത്തേണ്ടതാണ്.

സൂചനകൾ:
1. മൂല്യങ്ങൾ [0;30] എന്ന പരിധിയിലുള്ളവയായി പരിഗണിക്കാം
2. ഒരു ജീനായി, മൂല്യങ്ങളുടെ പട്ടിക ഉപയോഗിക്കാം
