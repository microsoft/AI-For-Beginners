# മൗണ്ടൻ കാർ രക്ഷപ്പെടാൻ പരിശീലനം

[AI for Beginners Curriculum](https://github.com/microsoft/ai-for-beginners) ലാബ് അസൈൻമെന്റ്.

OpenAI പരിസ്ഥിതിയിൽ [Mountain Car](https://www.gymlibrary.ml/environments/classic_control/mountain_car/) നിയന്ത്രിക്കാൻ RL ഏജന്റിനെ പരിശീലിപ്പിക്കുക എന്നതാണ് നിങ്ങളുടെ ലക്ഷ്യം.

പരിസ്ഥിതി സൃഷ്ടിക്കുന്നതിൽ നിന്ന് തുടങ്ങാം:


```python
import gym
env = gym.make('MountainCar-v0')
```

നമുക്ക് കാണാം ആ രാൻഡം പരീക്ഷണം എങ്ങനെ കാണപ്പെടുന്നു:


```python
state = env.reset()
while True:
    env.render()
    action = env.action_space.sample()
    state, reward, done, info = env.step(action)
    if done:
        break
```

ഇപ്പോൾ നോട്ട്‌ബുക്ക് മുഴുവനും നിങ്ങളുടെതാണ് - ഈ പ്രശ്നത്തിന് പാഠത്തിൽ നിന്നുള്ള പോളിസി ഗ്രാഡിയന്റുകളും ആക്ടർ-ക്രിട്ടിക് ആൽഗോരിതങ്ങളും സ്വീകാര്യമാക്കാൻ മടിക്കേണ്ട!
